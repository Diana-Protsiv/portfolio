# Приоритизация гипотез по увеличению выручки интернет-магазина и анализ A/B-теста

В основе исследования крупного интернет-магазина лежат гипотезы для увеличения его выручки.

**Цель исследования** — приоритизировать гипотезы по увеличению выручки интернет-магазина, запустить A/B-тест и проанализировать результаты.

**Описание данных**  

9 гипотез по увеличению выручки интернет-магазина с указанными параметрами Reach, Impact, Confidence, Effort.

Гипотезы:
    
- `Hypothesis` — краткое описание гипотезы

- `Reach` — охват пользователей по 10-балльной шкале

- `Impact` — влияние на пользователей по 10-балльной шкале

- `Confidence` — уверенность в гипотезе по 10-балльной шкале

- `Efforts` — затраты ресурсов на проверку гипотезы по 10-балльной шкале

Чем больше значение `Efforts`, тем дороже проверка гипотезы.

Заказы:
    
- `transactionId` — идентификатор заказа

- `visitorId` — идентификатор пользователя, совершившего заказ

- `date` — дата, когда был совершён заказ

- `revenue` — выручка заказа

- `group` — группа A/B-теста, в которую попал заказ (A или B)

Посетители:

- `date` — дата

- `group` — группа A/B-теста (A или B)

- `visitors` — количество пользователей в указанную дату в указанной группе A/B-теста

**План работы**

1. Загрузка и предобработка данных:
- загрузка данных о гипотезах, заказах и пользователях из CSV-файлов в переменные;
- вывод первых строк каждого набора данных;
- изучение общей информации о датафреймах;
- поиск пропущенных значений и дубликатов в данных;
- замена названий столбцов датафреймов;
- преобразование данных в нужные типы.

2. Приоритизация гипотез: 
- применение фреймворка ICE для приоритизации гипотез, сортировка их по убыванию приоритета;
- применение фреймворка RICE для приоритизации гипотез, сортировка их по убыванию приоритета;
- анализ изменения приоритизации гипотез при применении RICE вместо ICE.

3. Анализ A/B-теста:
- построение графика кумулятивной выручки по группам, написание выводов и предположений;
- построение графика кумулятивного среднего чека по группам, написание выводов и предположений;
- построение графика относительного изменения кумулятивного среднего чека группы B к группе A, написание выводов и предположений;
- построение графика кумулятивного среднего количества заказов на посетителя по группам, написание выводов и предположений;
- построение графика относительного изменения кумулятивного среднего количества заказов на посетителя группы B к группе A, написание выводов и предположений;
- построение точечного графика количества заказов по пользователям, написание выводов и предположений;
- подсчёт 95-го и 99-го перцентилей количества заказов на пользователя, выбор границы для определения аномальных пользователей;
- построение точечного графика стоимостей заказов, написание выводов и предположений;
- подсчёт 95-го и 99-го перцентилей стоимости заказов, выбор границы для определения аномальных заказов;
- подсчёт статистической значимости различий в среднем количестве заказов на посетителя между группами по «сырым» данным, написание выводов и предположений;
- подсчёт статистической значимости различий в среднем чеке заказа между группами по «сырым» данным, написание выводов и предположений;
- подсчёт статистической значимости различий в среднем количестве заказов на посетителя между группами по «очищенным» данным, написание выводов и предположений;
- подсчёт статистической значимости различий в среднем чеке заказа между группами по «очищенным» данным, написание выводов и предположений;
- принятие и объяснение решения по результатам теста, варианты решений:
   1. остановка теста, фиксация победы одной из групп;
   2. остановка теста, фиксация отсутствия различий между группами;
   3. продолжение теста.

4. Написание общих выводов по результатам исследования.

```python
ab = []
# создание пустого списка ‘ab’
for i in orders.query('group == "B"')['visitorId'].values:
    if i in orders.query('group == "A"')['visitorId'].values:
        ab.append(i)
# проверка каждого элемента из группы B на наличие в группе A,
# если совпадение найдено, то этот элемент добавляется в список ‘ab’
unique_ab = np.unique(ab)
# запись в ‘unique_ab’ уникальных элементов из списка ‘ab’
print('Количество уникальных пользователей, которые попали в группу A и группу B:', len(unique_ab), 'человек')
# подсчёт количества элементов в ‘unique_ab’
print('Общее количество уникальных пользователей:', orders['visitorId'].nunique(), 'человек')
# получение количества уникальных значений в столбце ‘visitorId’ датафрейма ‘orders’
print('Доля пользователей, попавших в обе группы, от общего количества уникальных пользователей:', 
      round(len(unique_ab) / orders['visitorId'].nunique() * 100, 1), '%')
# деление количества элементов в ‘unique_ab’ на количество уникальных значений в столбце ‘visitorId’
# умножение полученного значения на 100 и округление результата до 1 знака после запятой
# вывод на экран доли в процентах
```

Это небольшой процент, желательно было бы отфильтровать таких пользователей, но в имеющихся данных количество посетителей в день агрегировано. ## Приоритизация гипотез

Приоритизируем сформулированные гипотезы по увеличению выручки интернет-магазина, чтобы максимизировать ожидаемое улучшение ключевых метрик бизнеса.

### Приоритизация гипотез с применением фреймворка ICE

Применим фреймворк ICE (от англ. impact, confidence, efforts «влияние, уверенность, усилия») для приоритизации гипотез по увеличению выручки интернет-магазина, отсортируем их по убыванию приоритета.

Фреймворк ICE состоит из трёх компонентов:
- `impact` — насколько сильно изменение повлияет на пользователей, их опыт и удовлетворение от продукта;
- `confidence` — насколько мы уверены, что это изменение повлияет на пользователей именно так; 
- `efforts` — сколько стоит протестировать эту гипотезу.

Параметры фреймворка ICE оценены по шкале от 0 до 10.

```python
hypotheses['ice'] = round(hypotheses['impact'] * hypotheses['confidence'] / hypotheses['efforts'], 1)
# подсчёт ICE для гипотез с округлением до 1-го знака после запятой
# добавление полученных значений ICE к таблице ‘hypotheses’
ice = hypotheses[['hypothesis', 'ice']].sort_values(by='ice', ascending=False)
# запись в переменную ‘ice’ столбцов ‘hypothesis’ и ‘ice’ датафрейма ‘hypotheses’, 
# отсортированных по убыванию значений столбца ‘ice’
ice.sort_values(by='ice').plot(x='hypothesis', y='ice', kind='barh', figsize=(15, 4), legend=False)
# построение барного графика для значений столбца ‘ice’
plt.title('Приоритизация гипотез с применением фреймворка ICE', fontsize=12)
# установка заголовка для графика
plt.xlabel('ICE', fontsize=10)
# установка подписи для горизонтальной оси графика
plt.ylabel('Гипотеза', fontsize=10)
# установка подписи для вертикальной оси графика
plt.show()
# отображение графика
ice
# вывод таблицы ‘ice’, отсортированной по убыванию значений столбца ‘ice’
```

Наиболее перспективной по ICE (16.2) является 9-я гипотеза по увеличению выручки интернет-магазина «Запустить акцию, дающую скидку на товар в день рождения». Также к более приоритетным по ICE можно отнести 1-ю (13.3) и 8-ю (11.2) гипотезы: «Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей» и «Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок».

Самая неперспективная гипотеза по ICE (1.0) — это 5-я гипотеза «Изменить цвет фона главной страницы, чтобы увеличить вовлеченность пользователей».

### Приоритизация гипотез с применением фреймворка RICE

Теперь используем фреймворк RICE (R от англ. reach, «охват») для приоритизации гипотез по увеличению выручки интернет-магазина.

Фреймворк RICE состоит из четырёх компонентов:
- `reach` — скольких пользователей затронет изменение, которое мы хотим внести;
- `impact` — насколько сильно изменение повлияет на пользователей, их опыт и удовлетворение от продукта;
- `confidence` — насколько мы уверены, что это изменение повлияет на пользователей именно так; 
- `efforts` — сколько стоит протестировать эту гипотезу.

Как и в случае ICE, параметры фреймворка RICE оценены по шкале от 0 до 10.

```python
hypotheses['rice'] = hypotheses['reach'] * hypotheses['impact'] * hypotheses['confidence'] / hypotheses['efforts']
# подсчёт RICE для гипотез
# добавление полученных значений RICE к таблице ‘hypotheses’
rice = hypotheses[['hypothesis', 'rice']].sort_values(by='rice', ascending=False)
# запись в переменную ‘rice’ столбцов ‘hypothesis’ и ‘rice’ датафрейма ‘hypotheses’, 
# отсортированных по убыванию значений столбца ‘rice’
rice.sort_values(by='rice').plot(x='hypothesis', y='rice', kind='barh', figsize=(15, 4), legend=False)
# построение барного графика для значений столбца ‘rice’
plt.title('Приоритизация гипотез с применением фреймворка RICE', fontsize=12)
# установка заголовка для графика
plt.xlabel('RICE', fontsize=10)
# установка подписи для горизонтальной оси графика
plt.ylabel('Гипотеза', fontsize=10)
# установка подписи для вертикальной оси графика
plt.show()
# отображение графика
rice
# вывод таблицы ‘rice’, отсортированной по убыванию значений столбца ‘rice’
```

Добавление параметра, связанного с охватом пользователей, привело к изменению приоритетов гипотез по увеличению выручки интернет-магазина.

По RICE самая перспективная гипотеза (RICE равен 112) — это 8-я гипотеза «Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок». В 2 раза меньше RICE (равен 56) у 3-й гипотезы «Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа». Третье место по приоритету (RICE равен 40) занимают две гипотезы — 1-я «Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей» и 7-я «Показать на главной странице баннеры с актуальными акциями и распродажами, чтобы увеличить конверсию».

Лишь 16.2 составляет RICE у 9-й гипотезы «Запустить акцию, дающую скидку на товар в день рождения», что совпадает со значением ICE (равен 16.2). Это можно объяснить тем, что данная гипотеза затрагивает очень мало пользователей. Наименее перспективной по RICE (равен 3), как и по ICE (равен 1), является 5-я гипотеза «Изменить цвет фона главной страницы, чтобы увеличить вовлеченность пользователей».

Выведем список гипотез по увеличению выручки интернет-магазина вместе с ICE, RICE и их параметрами.

```python
hypotheses.sort_values(by='rice', ascending=False)
# вывод столбцов датафрейма ‘hypotheses’, 
# отсортированных по убыванию значений столбца ‘rice’
```

У наиболее приоритетной по RICE и входящей в топ-3 по приоритету по ICE гипотезы «Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок» охват пользователей оценён в 10 баллов из 10, влияние на пользователей — в 7 баллов, уверенность в гипотезе — в 8 баллов, затраты ресурсов на проверку гипотезы — в 5 баллов. Форма подписки может быть размещена на всех основных страницах сайта, что обеспечит максимальный охват пользователей. Могут быть получены целевые подписчики, заинтересованные в продуктах интернет-магазина. Они могут стать частью сообщества, что укрепит связь с брендом и увеличит лояльность. Рассылка новостей, статей и специальных предложений поможет поддерживать интерес аудитории и стимулировать её к совершению покупок.

У гипотезы «Запустить акцию, дающую скидку на товар в день рождения» параметр reach равен 1, поэтому он не оказал влияния на RICE. Причиной низкого охвата пользователей может являться то, что среди посетителей магазина мало тех, кто является именинником при покупке товаров, так как день рождения бывает только один раз в году. Соберём агрегированные кумулятивные по дням данные о заказах и посетителях интернет-магазина. Объявим датафрейм `cumulative_data` со столбцами:

- `date` — дата

- `group` — группа A/B-теста (A или B)

- `orders` — кумулятивное количество заказов на указанную дату в указанной группе

- `buyers` — кумулятивное количество пользователей, совершивших хотя бы один заказ, на указанную дату в указанной группе

- `revenue` — кумулятивная выручка на указанную дату в указанной группе

- `visitors` — кумулятивное количество посетителей интернет-магазина на указанную дату в определённой группе

```python
dates_groups = orders[['date', 'group']].drop_duplicates()
# создание массива уникальных пар значений дат и групп A/B-теста

orders_aggregated = dates_groups.apply(
    lambda x: orders[np.logical_and(
        orders['date'] <= x['date'], orders['group'] == x['group'])].agg(
        {'date' : 'max', 'group' : 'max', 'transaction_id' : 'nunique', 'visitor_id' : 'nunique', 'revenue' : 'sum'}), 
    axis=1).sort_values(by=['date', 'group'])
# получение строк таблицы ‘orders’, дата которых меньше или равна дате элемента из ‘dates_groups’, 
# а группа теста равна группе из ‘dates_groups’
# вычисление максимальной даты
# рассчёт максимума для группы, чтобы столбец ‘group’ не пропал из итогового вывода
# вычисление числа уникальных ID заказов и пользователей
# подсчёт суммы средних чеков
# применение методов к каждой строке датафрейма
# сортировка результатов по столбцам ‘date’ и ‘group’
# получение агрегированных кумулятивных по дням данных о заказах
 
visitors_aggregated = dates_groups.apply(
    lambda x: visitors[np.logical_and(
        visitors['date'] <= x['date'], visitors['group'] == x['group'])].agg(
        {'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1).sort_values(by=['date', 'group'])
# получение строк таблицы ‘visitors’, дата которых меньше или равна дате элемента из ‘dates_groups’, 
# а группа теста равна группе из ‘dates_groups’
# вычисление максимальной даты
# рассчёт максимума для группы, чтобы столбец ‘group’ не пропал из итогового вывода
# подсчёт суммарного количества пользователей
# применение методов к каждой строке датафрейма
# сортировка результатов по столбцам ‘date’ и ‘group’
# получение агрегированных кумулятивных по дням данных о посетителях интернет-магазина

cumulative_data = orders_aggregated.merge(visitors_aggregated, left_on=['date', 'group'], right_on=['date', 'group'])
# объединение кумулятивных данных в одной таблице ‘cumulative_data’
cumulative_data.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
# присваивание названий столбцам таблицы ‘cumulative_data’

cumulative_data
# вывод таблицы ‘cumulative_data’ на экран
```

Пользователи, принявшие участие в A/B-тесте, посещали интернет-магазин с 1 по 31 августа 2019 года. Соответственно, длительность эксперимента составила 1 месяц.

У группы B кумулятивное значение среднего чека колебалось с 1 по 18 августа в пределах 5-7 тыс. 19 августа средний чек, как и выручка, резко увеличился до 10 тыс. 214. ### Подсчёт статистической значимости различий в среднем количестве заказов на посетителя и в среднем чеке заказа между группами по «сырым» данным

Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным — без удаления аномальных пользователей.

Сформулируем гипотезы.

- Нулевая гипотеза: различий в среднем количестве заказов между группами нет. 
- Альтернативная гипотеза: различия в среднем количестве заказов между группами есть.

Создадим переменные `orders_by_visitors_a` и `orders_by_visitors_b` со столбцами `visitor_id`, `orders`. В них для пользователей, которые заказывали хотя бы 1 раз, укажем число совершённых заказов.

```python
orders_by_visitors_a = (
    orders[orders['group'] == 'A'].groupby('visitor_id', as_index=False).agg({'transaction_id': pd.Series.nunique}))
# выбор подмножества датафрейма ‘orders’, где значение столбца ‘group’ равно ‘A’
# группировка данных по столбцу ‘visitor_id’
# подсчёт уникальных значений столбца ‘transaction_id’ по сгрупированным данным
# сохранение результата в переменной ‘orders_by_visitors_a’

orders_by_visitors_a.columns = ['visitor_id', 'orders']
# присваивание названий столбцам таблицы ‘orders_by_visitors_a’

orders_by_visitors_b = (
    orders[orders['group'] == 'B'].groupby('visitor_id', as_index=False).agg({'transaction_id': pd.Series.nunique}))
# выбор подмножества датафрейма ‘orders’, где значение столбца ‘group’ равно ‘B’
# группировка данных по столбцу ‘visitor_id’
# подсчёт уникальных значений столбца ‘transaction_id’ по сгрупированным данным
# сохранение результата в переменной ‘orders_by_visitors_b’

orders_by_visitors_b.columns = ['visitor_id', 'orders']
# присваивание названий столбцам таблицы ‘orders_by_visitors_b’
```

Объявим переменные `sample_a` и `sample_b`, в которых пользователям из разных групп будет соответствовать количество заказов. Тем, кто ничего не заказал, будут соответствовать нули. Это нужно, чтобы подготовить выборки к проверке критерием Манна-Уитни.

```python
visitors_a_daily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
# создание датафрейма ‘visitors_a_daily’, содержащего только те строки из исходного датафрейма ‘visitors’, 
# где значение столбца group равно ‘A’
# выбор столбцов ‘date’ и ‘visitors’, которые будут сохранены в новом датафрейме
# получение количества пользователей в выбранную дату в группе A

visitors_a_daily.columns = ['date', 'visitors_per_date_a']
# присваивание названий столбцам датафрейма ‘visitors_a_daily’

sample_a = pd.concat(
    [orders_by_visitors_a['orders'], pd.Series(
        0, index=np.arange(
            visitors_a_daily['visitors_per_date_a'].sum() - len(orders_by_visitors_a['orders'])), name='orders')], axis=0)
# объединение 2-х объектов Series: кол-ва заказов для каждого из пользователей группы A и нулей для пользователей группы A, 
# которые ничего не заказывали
# количество нулей равно разнице между суммой посетителей и количеством записей о заказах
# передача аргумента, сообщающего, что объекты Series нужно объединить по строкам (то есть записать подряд)

visitors_b_daily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
# создание датафрейма ‘visitors_b_daily’, содержащего только те строки из исходного датафрейма ‘visitors’, 
# где значение столбца group равно ‘B’
# выбор столбцов ‘date’ и ‘visitors’, которые будут сохранены в новом датафрейме
# получение количества пользователей в выбранную дату в группе B

visitors_b_daily.columns = ['date', 'visitors_per_date_b']
# присваивание названий столбцам датафрейма ‘visitors_b_daily’

sample_b = pd.concat(
    [orders_by_visitors_b['orders'], pd.Series(
        0, index=np.arange(
            visitors_b_daily['visitors_per_date_b'].sum() - len(orders_by_visitors_b['orders'])), name='orders')], axis=0)
# объединение 2-х объектов Series: кол-ва заказов для каждого из пользователей группы B и нулей для пользователей группы B, 
# которые ничего не заказывали
# количество нулей равно разнице между суммой посетителей и количеством записей о заказах
# передача аргумента, сообщающего, что объекты Series нужно объединить по строкам (то есть записать подряд)
```

Посчитаем статистическую значимость различия среднего количества заказов на пользователя между группами по «сырым» данным. Применим тест Манна-Уитни. Выведем p-value для сравнения среднего между группами. Будем использовать значение 0.05 в качестве уровня статистической значимости.

Кроме того, вычислим и выведем среднее число заказов на пользователя группы A и группы B, а также относительное различие в среднем между группами A/B-теста по «сырым» данным.

```python
alpha = 0.05
# создание переменной ‘alpha’, равной 0.05, для установки уровня статистической значимости

pvalue = stats.mannwhitneyu(sample_a, sample_b)[1]
# подсчёт статистической значимости различия среднего количества заказов между группами
# применение теста Манна-Уитни

print(f'p-value: {"{0:.3f}".format(pvalue)}')
# вывод p-value для сравнения среднего между группами 
# округление p-value до 3-х знаков после запятой

if pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
# проверка, меньше ли p-значение результатов исследования, чем уровень значимости
# если это так, то нулевая гипотеза отвергается и выводится соответствующее сообщение
# в противном случае выводится ссобщение о том, что нет оснований отвергнуть нулевую гипотезу

print('Среднее число заказов на пользователя группы А по «сырым» данным:', "{:.3f}".format(sample_a.mean()))
# вычисление среднего значения в выборке ‘sample_a’
# округление до 3-х знаков после запятой

print('Среднее число заказов на пользователя группы B по «сырым» данным:', "{:.3f}".format(sample_b.mean()))
# вычисление среднего значения в выборке ‘sample_b’
# округление до 3-х знаков после запятой

print('Относительное различие среднего кол-ва заказов на посетителя между группами по «сырым» данным:', 
      "{0:.2%}".format(sample_b.mean() / sample_a.mean() - 1))
# вычисление и вывод относительного различия в среднем между группами
# округление до 2-х знаков после запятой
```

P-value составляет 0.017, что меньше заданного уровня значимости, равного 0.05, поэтому нулевую гипотезу отвергаем. По «сырым» данным можно сделать вывод, что в среднем количестве заказов на посетителя между группами A/B-теста есть статистически значимые различия. Относительный прирост среднего числа заказов группы B к конверсии группы A равен 13.8%.

Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным.

- Нулевая гипотеза: различий в среднем чеке между группами нет. 
- Альтернативная гипотеза: различия в среднем чеке между группами есть.

Чтобы рассчитать статистическую значимость различий в среднем чеке, передадим критерию Манна-Уитни данные о выручке с заказов. Будем использовать значение 0.05 в качестве уровня статистической значимости.

А ещё найдём средние чеки групп A и B, а также относительное различие в среднем чеке между группами A/B-теста по «сырым» данным.

```python
pvalue = stats.mannwhitneyu(orders[orders['group'] == 'A']['revenue'], orders[orders['group'] == 'B']['revenue'])[1]
# подсчёт статистической значимости различия средних чеков между группами
# применение теста Манна-Уитни

print(f'p-value: {"{0:.3f}".format(pvalue)}')
# вывод p-value для сравнения средних чеков между группами 
# округление p-value до 3-х знаков после запятой

if pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
# проверка, меньше ли p-значение результатов исследования, чем уровень значимости
# если это так, то нулевая гипотеза отвергается и выводится соответствующее сообщение
# в противном случае выводится ссобщение о том, что нет оснований отвергнуть нулевую гипотезу

print('Средний чек заказа группы A по «сырым» данным:', 
      "{0:.2f}".format(orders[orders['group'] == 'A']['revenue'].mean()))
# вывод среднего чека группы A
# округление результата до 2-х знаков после запятой

print('Средний чек заказа группы B по «сырым» данным:',
      "{0:.2f}".format(orders[orders['group'] == 'B']['revenue'].mean()))
# вывод среднего чека группы B
# округление результата до 2-х знаков после запятой

print('Относительное различие в среднем чеке между группами A/B-теста по «сырым» данным:',
    "{0:.2%}".format(
        orders[orders['group'] == 'B']['revenue'].mean() / orders[orders['group'] == 'A']['revenue'].mean() - 1))
# вывод относительного прироста среднего чека группы B
# округление результата до 2-х знаков после запятой
```

P-value составляет 0.729, что значительно больше заданного уровня значимости, равного 0.05. Значит, причин отвергать нулевую гипотезу и считать, что в среднем чеке между группами A/B-теста по «сырым» данным есть статистически значимые различия, нет. При этом средний чек группы B (9 тыс. 230) намного выше среднего чека группы A (7 тыс. 333). Относительное различие среднего чека между группами по «сырым» данным равно 25.9%.

### Подсчёт статистической значимости различий в среднем количестве заказов на посетителя и в среднем чеке заказа между группами по «очищенным» данным

Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя и в среднем чеке заказа между группами по «очищенным» данным.

Найденные ранее 95-й и 99-й перцентили средних чеков равны 28 тыс. и 58 тыс. 233. А 95-й и 99-й перцентили количества заказов на одного пользователя равны двум или четырём заказам на пользователя.

Мы приняли за аномальных пользователей тех, кто совершил 3 заказа или совершил заказ дороже 30 тыс. Так мы уберём 5% пользователей с наибольшим числом заказов и от 1% до 5% пользователей с дорогими заказами.

Сделаем срезы пользователей с числом заказов больше 2 — `visitors_with_many_orders` и пользователей, совершивших заказы дороже 30 000 — `visitors_with_expensive_orders`. Объединим их в таблице `abnormal_visitors`.

```python
# number_of_orders_limit = np.percentile(orders_by_visitors['orders'], 99)
# вычисление 99-го процентиля значений столбца ‘orders’ в датафрейме ‘orders_by_visitors’
# задание границы количества заказов

number_of_orders_limit = np.percentile(orders_by_visitors['orders'], 95)
# вычисление 95-го процентиля значений столбца ‘orders’ в датафрейме ‘orders_by_visitors’
# задание границы количества заказов

visitors_with_many_orders = pd.concat(
    [orders_by_visitors_a[orders_by_visitors_a['orders'] > number_of_orders_limit]['visitor_id'],
        orders_by_visitors_b[orders_by_visitors_b['orders'] > number_of_orders_limit]['visitor_id'],],
    axis=0)
# объединение посетителей, 
# у которых количество заказов превышает заданное ограничение

# order_value_limit = round(np.percentile(orders['revenue'], 95.905))
# вычисление 95.905-го процентиля значений столбца ‘revenue’ в датафрейме ‘orders’
# задание границы стоимости заказа

order_value_limit = round(np.percentile(orders['revenue'], 95.5214))
# вычисление 95.5214-го процентиля значений столбца ‘revenue’ в датафрейме ‘orders’
# задание границы стоимости заказа

visitors_with_expensive_orders = orders[orders['revenue'] > order_value_limit]['visitor_id']
# отбор посетителей, у которых есть дорогие заказы (где выручка больше заданного ограничения)

abnormal_visitors = (pd.concat([visitors_with_many_orders, visitors_with_expensive_orders], axis=0)
    .drop_duplicates()
    .sort_values())
# объединение в таблице ‘abnormal_visitors’ пользователей с числом заказов больше 2 
# и пользователей, совершивших заказы дороже 30 000
# удаление дубликатов, сортировка значений по возрастанию

abnormal_visitors.head(5).to_frame()
# вывод первых 5-ти строк переменной ‘abnormal_visitors’
```

Узнаем, сколько всего аномальных пользователей.

```python
print('Количество аномальных пользователей:', abnormal_visitors.shape[0], 'человек')
# получение количества строк в таблице ‘abnormal_visitors’
```

Всего 68 аномальных пользователей. Узнаем, как их действия повлияли на результаты теста. Посчитаем статистическую значимость различий в среднем количестве заказов между группами теста по «очищенным» данным — после удаления аномальных пользователей.

- Нулевая гипотеза: различий в среднем количестве заказов между группами нет.
- Альтернативная гипотеза: различия в среднем количестве заказов между группами есть.

Сначала подготовим выборки количества заказов по пользователям по группам A/B-теста.

```python
sample_a_filtered = pd.concat(
    [
        orders_by_visitors_a[
            np.logical_not(orders_by_visitors_a['visitor_id'].isin(abnormal_visitors))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitors_a_daily['visitors_per_date_a'].sum() - len(orders_by_visitors_a['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0)
# объединение 2-х объектов Series: кол-ва заказов для каждого из пользователей группы A и нулей для пользователей группы A, 
# которые ничего не заказывали 
# количество нулей равно разнице между суммой посетителей и количеством записей о заказах
# с исключением аномальных пользователей

sample_b_filtered = pd.concat(
    [
        orders_by_visitors_b[
            np.logical_not(orders_by_visitors_b['visitor_id'].isin(abnormal_visitors))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitors_b_daily['visitors_per_date_b'].sum() - len(orders_by_visitors_b['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0)
# объединение 2-х объектов Series: кол-ва заказов для каждого из пользователей группы B и нулей для пользователей группы B, 
# которые ничего не заказывали 
# количество нулей равно разнице между суммой посетителей и количеством записей о заказах
# с исключением аномальных пользователей
```

Применим статистический критерий Манна-Уитни к полученным выборкам `sample_a_filtered` и `sample_b_filtered`.

Будем использовать значение 0.05 в качестве уровня статистической значимости.

```python
pvalue = stats.mannwhitneyu(sample_a_filtered, sample_b_filtered)[1]
# подсчёт статистической значимости различия среднего количества заказов между группами
# применение теста Манна-Уитни

print(f'p-value: {"{0:.3f}".format(pvalue)}')
# вывод p-value для сравнения среднего между группами 
# округление p-value до 3-х знаков после запятой

if pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
# проверка, меньше ли p-значение результатов исследования, чем уровень значимости
# если это так, то нулевая гипотеза отвергается и выводится соответствующее сообщение
# в противном случае выводится ссобщение о том, что нет оснований отвергнуть нулевую гипотезу

print('Среднее число заказов на пользователя группы А по «очищенным» данным:', "{:.3f}".format(sample_a_filtered.mean()))
# вычисление среднего значения в выборке ‘sample_a_filtered’
# округление до 3-х знаков после запятой

print('Среднее число заказов на пользователя группы B по «очищенным» данным:', "{:.3f}".format(sample_b_filtered.mean()))
# вычисление среднего значения в выборке ‘sample_b_filtered’
# округление до 3-х знаков после запятой

print('Относительное различие среднего кол-ва заказов на посетителя между группами по «очищенным» данным:', 
      "{0:.2%}".format(sample_b_filtered.mean() / sample_a_filtered.mean() - 1))
# вычисление и вывод относительного различия в среднем между группами
# округление до 2-х знаков после запятой
```

Как и в случае с «сырыми» данными, статистическая значимость достигнута. Группа B лучше группы A.

P-value составляет 0.013, что меньше заданного уровня значимости, равного 0.05. Значит, нулевую гипотезу о том, что статистически значимых различий в среднем числе заказов между группами нет, отвергаем. По «очищенным» данным можно сделать вывод, что в среднем количестве заказов на посетителя между группами A/B-теста есть статистически значимые различия. Относительный выигрыш группы B по «очищенным» данным равен 17.2%, что на 3.4% больше, чем по «сырым» данным.

Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «очищенным» данным. Будем использовать значение 0.05 в качестве уровня статистической значимости.

- Нулевая гипотеза: различий в среднем чеке между группами нет. 
- Альтернативная гипотеза: различия в среднем чеке между группами есть.

```python
pvalue = stats.mannwhitneyu(orders[
                np.logical_and(
                    orders['group'] == 'A',
                    np.logical_not(orders['visitor_id'].isin(abnormal_visitors)),
                )
            ]['revenue'],
            orders[
                np.logical_and(
                    orders['group'] == 'B',
                    np.logical_not(orders['visitor_id'].isin(abnormal_visitors)),
                )
            ]['revenue'])[1]
# подсчёт статистической значимости различия средних чеков между группами
# применение теста Манна-Уитни

print(f'p-value: {"{0:.3f}".format(pvalue)}')
# вывод p-value для сравнения средних чеков между группами 
# округление p-value до 3-х знаков после запятой

if pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
# проверка, меньше ли p-значение результатов исследования, чем уровень значимости
# если это так, то нулевая гипотеза отвергается и выводится соответствующее сообщение
# в противном случае выводится ссобщение о том, что нет оснований отвергнуть нулевую гипотезу

print('Средний чек заказа группы A по «очищенным» данным:', 
      "{0:.2f}".format(orders[
            np.logical_and(
                orders['group'] == 'A',
                np.logical_not(orders['visitor_id'].isin(abnormal_visitors)),
            )
        ]['revenue'].mean()))
# вывод среднего чека группы A
# округление результата до 2-х знаков после запятой

print('Средний чек заказа группы B по «очищенным» данным:',
      "{0:.2f}".format(orders[
            np.logical_and(
                orders['group'] == 'B',
                np.logical_not(orders['visitor_id'].isin(abnormal_visitors)),
            )
        ]['revenue'].mean()))
# вывод среднего чека группы B
# округление результата до 2-х знаков после запятой

print('Относительное различие в среднем чеке между группами A/B-теста по «очищенным» данным:',
    "{0:.2%}".format(
        orders[
            np.logical_and(
                orders['group'] == 'B',
                np.logical_not(orders['visitor_id'].isin(abnormal_visitors)),
            )
        ]['revenue'].mean()
        / orders[
            np.logical_and(
                orders['group'] == 'A',
                np.logical_not(orders['visitor_id'].isin(abnormal_visitors)),
            )
        ]['revenue'].mean()
        - 1))
# вывод относительного прироста среднего чека группы B
# округление результата до 2-х знаков после запятой
```

По «очищенным» данным p-value меньше на 0.009, чем p-value по «сырым» данным, и составляет 0.72, что всё равно намного больше заданного уровня значимости, равного 0.05. В связи с этим, нет оснований отвергать нулевую гипотезу и считать, что в среднем чеке между группами A/B-теста по «очищенным» данным есть статистически значимые различия. По данным без аномальных пользователей средний чек группы B (5 тыс. 192) чуть ниже среднего чека группы A (5 тыс. 315). Относительное различие среднего чека между группами по «очищенным» данным равно -2.31%.
