### –ò–∑—É—á–µ–Ω–∏–µ –æ–±—â–µ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –∫–∞–∂–¥–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ. –ü–æ–∏—Å–∫ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤

–í—ã–≤–µ–¥–µ–º –æ—Å–Ω–æ–≤–Ω—É—é –∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞—Ö `hypotheses`, `orders` –∏ `visitors`, —á–∏—Å–ª–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤, –∞ —Ç–∞–∫–∂–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –¥–æ–ª—é –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –≤—Å–µ—Ö —Å—Ç–æ–ª–±—Ü–æ–≤.

```python
def df_info(df, name):
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–≤–æ–¥–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
    print('\033[1m', f'–¢–∞–±–ª–∏—Ü–∞ ¬´{name}¬ª', '\033[0m')
    # –≤—ã–≤–æ–¥ –Ω–∞–∑–≤–∞–Ω–∏—è –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
    print(df.info())
    # –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Å—Ç—Ä—É–∫—Ç—É—Ä–µ –¥–∞–Ω–Ω—ã—Ö, —Ö—Ä–∞–Ω—è—â–∏—Ö—Å—è –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äòdf‚Äô
    print('-' * 65)
    # –≤—ã–≤–æ–¥ —Å–∏–º–≤–æ–ª–∞ ¬´-¬ª 65 —Ä–∞–∑
    print(df.describe())
    # –≤—ã–≤–æ–¥ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –∑–Ω–∞—á–µ–Ω–∏—è—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòdf‚Äô
    print('-' * 65)
    # –≤—ã–≤–æ–¥ —Å–∏–º–≤–æ–ª–∞ ¬´-¬ª 65 —Ä–∞–∑
    print('\033[1m', f'–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ —Ç–∞–±–ª–∏—Ü–µ ¬´{name}¬ª:', df.duplicated().sum(), '\033[0m')
    # –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ df‚Äô
    print('-' * 65)
    # –≤—ã–≤–æ–¥ —Å–∏–º–≤–æ–ª–∞ ¬´-¬ª 65 —Ä–∞–∑
    for i in df.columns:
        print('\033[1m', f'–°—Ç–æ–ª–±–µ—Ü ¬´{i}¬ª -', '–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ NaN:', df[i].isna().sum(),
        ', –ø—Ä–æ—Ü–µ–Ω—Ç NaN:', round(df[i].isna().mean() * 100, 2), '%', '\033[0m')
        # –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∏ –¥–æ–ª–∏ –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—â–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π (NaN) –≤ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòdf‚Äô
        print('-' * 65)
        # –≤—ã–≤–æ–¥ —Å–∏–º–≤–æ–ª–∞ ¬´-¬ª 65 —Ä–∞–∑
    
df_info(hypotheses, 'hypotheses')
# –≤—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ df_info –¥–ª—è –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòhypotheses‚Äô
df_info(orders, 'orders')
# –≤—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ df_info –¥–ª—è –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòorders‚Äô
df_info(visitors, 'visitors')
# –≤—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ df_info –¥–ª—è –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòvisitors‚Äô
```

```python
ab = []
# —Å–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ —Å–ø–∏—Å–∫–∞ ‚Äòab‚Äô
for i in orders.query('group == "B"')['visitorId'].values:
    if i in orders.query('group == "A"')['visitorId'].values:
        ab.append(i)
# –ø—Ä–æ–≤–µ—Ä–∫–∞ –∫–∞–∂–¥–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –∏–∑ –≥—Ä—É–ø–ø—ã B –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –≤ –≥—Ä—É–ø–ø–µ A,
# –µ—Å–ª–∏ —Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ –Ω–∞–π–¥–µ–Ω–æ, —Ç–æ —ç—Ç–æ—Ç —ç–ª–µ–º–µ–Ω—Ç –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è –≤ —Å–ø–∏—Å–æ–∫ ‚Äòab‚Äô
unique_ab = np.unique(ab)
# –∑–∞–ø–∏—Å—å –≤ ‚Äòunique_ab‚Äô —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ‚Äòab‚Äô
print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–ø–∞–ª–∏ –≤ –≥—Ä—É–ø–ø—É A –∏ –≥—Ä—É–ø–ø—É B:', len(unique_ab), '—á–µ–ª–æ–≤–µ–∫')
# –ø–æ–¥—Å—á—ë—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ ‚Äòunique_ab‚Äô
print('–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π:', orders['visitorId'].nunique(), '—á–µ–ª–æ–≤–µ–∫')
# –ø–æ–ª—É—á–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ ‚ÄòvisitorId‚Äô –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòorders‚Äô
print('–î–æ–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –ø–æ–ø–∞–≤—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã, –æ—Ç –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π:', 
      round(len(unique_ab) / orders['visitorId'].nunique() * 100, 1), '%')
# –¥–µ–ª–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ ‚Äòunique_ab‚Äô –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ ‚ÄòvisitorId‚Äô
# —É–º–Ω–æ–∂–µ–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ 100 –∏ –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –¥–æ 1 –∑–Ω–∞–∫–∞ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
# –≤—ã–≤–æ–¥ –Ω–∞ —ç–∫—Ä–∞–Ω –¥–æ–ª–∏ –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö
```

```python
hypotheses['ice'] = round(hypotheses['impact'] * hypotheses['confidence'] / hypotheses['efforts'], 1)
# –ø–æ–¥—Å—á—ë—Ç ICE –¥–ª—è –≥–∏–ø–æ—Ç–µ–∑ —Å –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ–º –¥–æ 1-–≥–æ –∑–Ω–∞–∫–∞ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π ICE –∫ —Ç–∞–±–ª–∏—Ü–µ ‚Äòhypotheses‚Äô
ice = hypotheses[['hypothesis', 'ice']].sort_values(by='ice', ascending=False)
# –∑–∞–ø–∏—Å—å –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é ‚Äòice‚Äô —Å—Ç–æ–ª–±—Ü–æ–≤ ‚Äòhypothesis‚Äô –∏ ‚Äòice‚Äô –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòhypotheses‚Äô, 
# –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòice‚Äô
ice.sort_values(by='ice').plot(x='hypothesis', y='ice', kind='barh', figsize=(15, 4), legend=False)
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –±–∞—Ä–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòice‚Äô
plt.title('–ü—Ä–∏–æ—Ä–∏—Ç–∏–∑–∞—Ü–∏—è –≥–∏–ø–æ—Ç–µ–∑ —Å –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º —Ñ—Ä–µ–π–º–≤–æ—Ä–∫–∞ ICE', fontsize=12)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
plt.xlabel('ICE', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞
plt.ylabel('–ì–∏–ø–æ—Ç–µ–∑–∞', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞
ice
# –≤—ã–≤–æ–¥ —Ç–∞–±–ª–∏—Ü—ã ‚Äòice‚Äô, –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòice‚Äô
```

- `impact` ‚Äî –Ω–∞—Å–∫–æ–ª—å–∫–æ —Å–∏–ª—å–Ω–æ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –ø–æ–≤–ª–∏—è–µ—Ç –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∏—Ö –æ–ø—ã—Ç –∏ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–∏–µ –æ—Ç –ø—Ä–æ–¥—É–∫—Ç–∞;

- `efforts` ‚Äî —Å–∫–æ–ª—å–∫–æ —Å—Ç–æ–∏—Ç –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å —ç—Ç—É –≥–∏–ø–æ—Ç–µ–∑—É.

–ö–∞–∫ –∏ –≤ —Å–ª—É—á–∞–µ ICE, –ø–∞—Ä–∞–º–µ—Ç—Ä—ã —Ñ—Ä–µ–π–º–≤–æ—Ä–∫–∞ RICE –æ—Ü–µ–Ω–µ–Ω—ã –ø–æ —à–∫–∞–ª–µ –æ—Ç 0 –¥–æ 10.

```python
hypotheses['rice'] = hypotheses['reach'] * hypotheses['impact'] * hypotheses['confidence'] / hypotheses['efforts']
# –ø–æ–¥—Å—á—ë—Ç RICE –¥–ª—è –≥–∏–ø–æ—Ç–µ–∑
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π RICE –∫ —Ç–∞–±–ª–∏—Ü–µ ‚Äòhypotheses‚Äô
rice = hypotheses[['hypothesis', 'rice']].sort_values(by='rice', ascending=False)
# –∑–∞–ø–∏—Å—å –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é ‚Äòrice‚Äô —Å—Ç–æ–ª–±—Ü–æ–≤ ‚Äòhypothesis‚Äô –∏ ‚Äòrice‚Äô –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòhypotheses‚Äô, 
# –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòrice‚Äô
rice.sort_values(by='rice').plot(x='hypothesis', y='rice', kind='barh', figsize=(15, 4), legend=False)
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –±–∞—Ä–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòrice‚Äô
plt.title('–ü—Ä–∏–æ—Ä–∏—Ç–∏–∑–∞—Ü–∏—è –≥–∏–ø–æ—Ç–µ–∑ —Å –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º —Ñ—Ä–µ–π–º–≤–æ—Ä–∫–∞ RICE', fontsize=12)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
plt.xlabel('RICE', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞
plt.ylabel('–ì–∏–ø–æ—Ç–µ–∑–∞', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞
rice
# –≤—ã–≤–æ–¥ —Ç–∞–±–ª–∏—Ü—ã ‚Äòrice‚Äô, –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòrice‚Äô
```

–î–æ–±–∞–≤–ª–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞, —Å–≤—è–∑–∞–Ω–Ω–æ–≥–æ —Å –æ—Ö–≤–∞—Ç–æ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –ø—Ä–∏–≤–µ–ª–æ –∫ –∏–∑–º–µ–Ω–µ–Ω–∏—é –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–æ–≤ –≥–∏–ø–æ—Ç–µ–∑ –ø–æ —É–≤–µ–ª–∏—á–µ–Ω–∏—é –≤—ã—Ä—É—á–∫–∏ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-–º–∞–≥–∞–∑–∏–Ω–∞.

–ü–æ RICE —Å–∞–º–∞—è –ø–µ—Ä—Å–ø–µ–∫—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (RICE —Ä–∞–≤–µ–Ω 112) ‚Äî —ç—Ç–æ 8-—è –≥–∏–ø–æ—Ç–µ–∑–∞ ¬´–î–æ–±–∞–≤–∏—Ç—å —Ñ–æ—Ä–º—É –ø–æ–¥–ø–∏—Å–∫–∏ –Ω–∞ –≤—Å–µ –æ—Å–Ω–æ–≤–Ω—ã–µ —Å—Ç—Ä–∞–Ω–∏—Ü—ã, —á—Ç–æ–±—ã —Å–æ–±—Ä–∞—Ç—å –±–∞–∑—É –∫–ª–∏–µ–Ω—Ç–æ–≤ –¥–ª—è email-—Ä–∞—Å—Å—ã–ª–æ–∫¬ª. –í 2 —Ä–∞–∑–∞ –º–µ–Ω—å—à–µ RICE (—Ä–∞–≤–µ–Ω 56) —É 3-–π –≥–∏–ø–æ—Ç–µ–∑—ã ¬´–î–æ–±–∞–≤–∏—Ç—å –±–ª–æ–∫–∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–π —Ç–æ–≤–∞—Ä–æ–≤ –Ω–∞ —Å–∞–π—Ç –∏–Ω—Ç–µ—Ä–Ω–µ—Ç –º–∞–≥–∞–∑–∏–Ω–∞, —á—Ç–æ–±—ã –ø–æ–≤—ã—Å–∏—Ç—å –∫–æ–Ω–≤–µ—Ä—Å–∏—é –∏ —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –∑–∞–∫–∞–∑–∞¬ª. –¢—Ä–µ—Ç—å–µ –º–µ—Å—Ç–æ –ø–æ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç—É (RICE —Ä–∞–≤–µ–Ω 40) –∑–∞–Ω–∏–º–∞—é—Ç –¥–≤–µ –≥–∏–ø–æ—Ç–µ–∑—ã ‚Äî 1-—è ¬´–î–æ–±–∞–≤–∏—Ç—å –¥–≤–∞ –Ω–æ–≤—ã—Ö –∫–∞–Ω–∞–ª–∞ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è —Ç—Ä–∞—Ñ–∏–∫–∞, —á—Ç–æ –ø–æ–∑–≤–æ–ª–∏—Ç –ø—Ä–∏–≤–ª–µ–∫–∞—Ç—å –Ω–∞ 30% –±–æ–ª—å—à–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π¬ª –∏ 7-—è ¬´–ü–æ–∫–∞–∑–∞—Ç—å –Ω–∞ –≥–ª–∞–≤–Ω–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü–µ –±–∞–Ω–Ω–µ—Ä—ã —Å –∞–∫—Ç—É–∞–ª—å–Ω—ã–º–∏ –∞–∫—Ü–∏—è–º–∏ –∏ —Ä–∞—Å–ø—Ä–æ–¥–∞–∂–∞–º–∏, —á—Ç–æ–±—ã —É–≤–µ–ª–∏—á–∏—Ç—å –∫–æ–Ω–≤–µ—Ä—Å–∏—é¬ª.

–õ–∏—à—å 16.2 —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç RICE —É 9-–π –≥–∏–ø–æ—Ç–µ–∑—ã ¬´–ó–∞–ø—É—Å—Ç–∏—Ç—å –∞–∫—Ü–∏—é, –¥–∞—é—â—É—é —Å–∫–∏–¥–∫—É –Ω–∞ —Ç–æ–≤–∞—Ä –≤ –¥–µ–Ω—å —Ä–æ–∂–¥–µ–Ω–∏—è¬ª, —á—Ç–æ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º ICE (—Ä–∞–≤–µ–Ω 16.2). –≠—Ç–æ –º–æ–∂–Ω–æ –æ–±—ä—è—Å–Ω–∏—Ç—å —Ç–µ–º, —á—Ç–æ –¥–∞–Ω–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –∑–∞—Ç—Ä–∞–≥–∏–≤–∞–µ—Ç –æ—á–µ–Ω—å –º–∞–ª–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π. –ù–∞–∏–º–µ–Ω–µ–µ –ø–µ—Ä—Å–ø–µ–∫—Ç–∏–≤–Ω–æ–π –ø–æ RICE (—Ä–∞–≤–µ–Ω 3), –∫–∞–∫ –∏ –ø–æ ICE (—Ä–∞–≤–µ–Ω 1), —è–≤–ª—è–µ—Ç—Å—è 5-—è –≥–∏–ø–æ—Ç–µ–∑–∞ ¬´–ò–∑–º–µ–Ω–∏—Ç—å —Ü–≤–µ—Ç —Ñ–æ–Ω–∞ –≥–ª–∞–≤–Ω–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã, —á—Ç–æ–±—ã —É–≤–µ–ª–∏—á–∏—Ç—å –≤–æ–≤–ª–µ—á–µ–Ω–Ω–æ—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π¬ª.

–°–æ–±–µ—Ä—ë–º –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–µ –ø–æ –¥–Ω—è–º –¥–∞–Ω–Ω—ã–µ –æ –∑–∞–∫–∞–∑–∞—Ö –∏ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è—Ö –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-–º–∞–≥–∞–∑–∏–Ω–∞. –û–±—ä—è–≤–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º `cumulative_data` —Å–æ —Å—Ç–æ–ª–±—Ü–∞–º–∏:

- `date` ‚Äî –¥–∞—Ç–∞

- `group` ‚Äî –≥—Ä—É–ø–ø–∞ A/B-—Ç–µ—Å—Ç–∞ (A –∏–ª–∏ B)

- `orders` ‚Äî –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω—É—é –¥–∞—Ç—É –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–π –≥—Ä—É–ø–ø–µ

- `buyers` ‚Äî –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, —Å–æ–≤–µ—Ä—à–∏–≤—à–∏—Ö —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω –∑–∞–∫–∞–∑, –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω—É—é –¥–∞—Ç—É –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–π –≥—Ä—É–ø–ø–µ

- `revenue` ‚Äî –∫—É–º—É–ª—è—Ç–∏–≤–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω—É—é –¥–∞—Ç—É –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–π –≥—Ä—É–ø–ø–µ

- `visitors` ‚Äî –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª–µ–π –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-–º–∞–≥–∞–∑–∏–Ω–∞ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω—É—é –¥–∞—Ç—É –≤ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–π –≥—Ä—É–ø–ø–µ

```python
dates_groups = orders[['date', 'group']].drop_duplicates()
# —Å–æ–∑–¥–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–∞—Ä –∑–Ω–∞—á–µ–Ω–∏–π –¥–∞—Ç –∏ –≥—Ä—É–ø–ø A/B-—Ç–µ—Å—Ç–∞

orders_aggregated = dates_groups.apply(
    lambda x: orders[np.logical_and(
        orders['date'] <= x['date'], orders['group'] == x['group'])].agg(
        {'date' : 'max', 'group' : 'max', 'transaction_id' : 'nunique', 'visitor_id' : 'nunique', 'revenue' : 'sum'}), 
    axis=1).sort_values(by=['date', 'group'])
# –ø–æ–ª—É—á–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫ —Ç–∞–±–ª–∏—Ü—ã ‚Äòorders‚Äô, –¥–∞—Ç–∞ –∫–æ—Ç–æ—Ä—ã—Ö –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–∞ –¥–∞—Ç–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –∏–∑ ‚Äòdates_groups‚Äô, 
# –∞ –≥—Ä—É–ø–ø–∞ —Ç–µ—Å—Ç–∞ —Ä–∞–≤–Ω–∞ –≥—Ä—É–ø–ø–µ –∏–∑ ‚Äòdates_groups‚Äô
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –¥–∞—Ç—ã
# —Ä–∞—Å—Å—á—ë—Ç –º–∞–∫—Å–∏–º—É–º–∞ –¥–ª—è –≥—Ä—É–ø–ø—ã, —á—Ç–æ–±—ã —Å—Ç–æ–ª–±–µ—Ü ‚Äògroup‚Äô –Ω–µ –ø—Ä–æ–ø–∞–ª –∏–∑ –∏—Ç–æ–≥–æ–≤–æ–≥–æ –≤—ã–≤–æ–¥–∞
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö ID –∑–∞–∫–∞–∑–æ–≤ –∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
# –ø–æ–¥—Å—á—ë—Ç —Å—É–º–º—ã —Å—Ä–µ–¥–Ω–∏—Ö —á–µ–∫–æ–≤
# –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤ –∫ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
# —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º ‚Äòdate‚Äô –∏ ‚Äògroup‚Äô
# –ø–æ–ª—É—á–µ–Ω–∏–µ –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã—Ö –ø–æ –¥–Ω—è–º –¥–∞–Ω–Ω—ã—Ö –æ –∑–∞–∫–∞–∑–∞—Ö
 
visitors_aggregated = dates_groups.apply(
    lambda x: visitors[np.logical_and(
        visitors['date'] <= x['date'], visitors['group'] == x['group'])].agg(
        {'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1).sort_values(by=['date', 'group'])
# –ø–æ–ª—É—á–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫ —Ç–∞–±–ª–∏—Ü—ã ‚Äòvisitors‚Äô, –¥–∞—Ç–∞ –∫–æ—Ç–æ—Ä—ã—Ö –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–∞ –¥–∞—Ç–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –∏–∑ ‚Äòdates_groups‚Äô, 
# –∞ –≥—Ä—É–ø–ø–∞ —Ç–µ—Å—Ç–∞ —Ä–∞–≤–Ω–∞ –≥—Ä—É–ø–ø–µ –∏–∑ ‚Äòdates_groups‚Äô
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –¥–∞—Ç—ã
# —Ä–∞—Å—Å—á—ë—Ç –º–∞–∫—Å–∏–º—É–º–∞ –¥–ª—è –≥—Ä—É–ø–ø—ã, —á—Ç–æ–±—ã —Å—Ç–æ–ª–±–µ—Ü ‚Äògroup‚Äô –Ω–µ –ø—Ä–æ–ø–∞–ª –∏–∑ –∏—Ç–æ–≥–æ–≤–æ–≥–æ –≤—ã–≤–æ–¥–∞
# –ø–æ–¥—Å—á—ë—Ç —Å—É–º–º–∞—Ä–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
# –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤ –∫ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
# —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º ‚Äòdate‚Äô –∏ ‚Äògroup‚Äô
# –ø–æ–ª—É—á–µ–Ω–∏–µ –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã—Ö –ø–æ –¥–Ω—è–º –¥–∞–Ω–Ω—ã—Ö –æ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è—Ö –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-–º–∞–≥–∞–∑–∏–Ω–∞

cumulative_data = orders_aggregated.merge(visitors_aggregated, left_on=['date', 'group'], right_on=['date', 'group'])
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –≤ –æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ ‚Äòcumulative_data‚Äô
cumulative_data.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
# –ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞–º —Ç–∞–±–ª–∏—Ü—ã ‚Äòcumulative_data‚Äô

cumulative_data
# –≤—ã–≤–æ–¥ —Ç–∞–±–ª–∏—Ü—ã ‚Äòcumulative_data‚Äô –Ω–∞ —ç–∫—Ä–∞–Ω
```

–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏, –ø—Ä–∏–Ω—è–≤—à–∏–µ —É—á–∞—Å—Ç–∏–µ –≤ A/B-—Ç–µ—Å—Ç–µ, –ø–æ—Å–µ—â–∞–ª–∏ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-–º–∞–≥–∞–∑–∏–Ω —Å 1 –ø–æ 31 –∞–≤–≥—É—Å—Ç–∞ 2019 –≥–æ–¥–∞. –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ —Å–æ—Å—Ç–∞–≤–∏–ª–∞ 1 –º–µ—Å—è—Ü.

### –ê–Ω–∞–ª–∏–∑ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –∏ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –ø–æ –≥—Ä—É–ø–ø–∞–º, –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A

–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ –¥–Ω—è–º –∏ –≥—Ä—É–ø–ø–∞–º A/B-—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è.

```python
cumulative_revenue_a = cumulative_data[cumulative_data['group'] == 'A'][['date', 'revenue', 'orders']]
# —Å–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–∫–∞–∑–æ–≤ –∏ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–æ–π –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ –ê

cumulative_revenue_b = cumulative_data[cumulative_data['group'] == 'B'][['date', 'revenue', 'orders']]
# —Å–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–∫–∞–∑–æ–≤ –∏ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–æ–π –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ B

plt.figure(figsize=(18, 4))
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ —Ä–∞–∑–º–µ—Ä–∞ –≥—Ä–∞—Ñ–∏–∫–∞

plt.plot(cumulative_revenue_a['date'], cumulative_revenue_a['revenue'], label='A')
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –≤—ã—Ä—É—á–∫–∏ –≥—Ä—É–ø–ø—ã –ê

plt.plot(cumulative_revenue_b['date'], cumulative_revenue_b['revenue'], label='B')
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –≤—ã—Ä—É—á–∫–∏ –≥—Ä—É–ø–ø—ã B

plt.title('–ö—É–º—É–ª—è—Ç–∏–≤–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –ø–æ –¥–Ω—è–º –∏ –≥—Ä—É–ø–ø–∞–º A/B-—Ç–µ—Å—Ç–∞', fontsize=12)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞

plt.xlabel('–î–∞—Ç–∞', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞

plt.ylabel('–ö—É–º—É–ª—è—Ç–∏–≤–Ω–∞—è –≤—ã—Ä—É—á–∫–∞', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞

plt.legend();
# –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ª–µ–≥–µ–Ω–¥—ã –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ
```

```python
cumulative_data['conversion'] = cumulative_data['orders'] / cumulative_data['visitors']
# —Ä–∞—Å—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤ —Ç–∞–±–ª–∏—Ü—É ‚Äòcumulative_data‚Äô —Å—Ç–æ–ª–±—Ü–∞ ‚Äòconversion‚Äô 
# c –æ—Ç–Ω–æ—à–µ–Ω–∏–µ–º —á–∏—Å–ª–∞ –∑–∞–∫–∞–∑–æ–≤ –∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–π –≥—Ä—É–ø–ø–µ –≤ —É–∫–∞–∑–∞–Ω–Ω—ã–π –¥–µ–Ω—å

cumulative_data_a = cumulative_data[cumulative_data['group'] == 'A']
# —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π cumulative_data_a –¥–∞–Ω–Ω—ã—Ö –æ –∑–∞–∫–∞–∑–∞—Ö –≤ –≥—Ä—É–ø–ø–µ A

cumulative_data_b = cumulative_data[cumulative_data['group'] == 'B'] 
# —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π cumulative_data_b –¥–∞–Ω–Ω—ã—Ö –æ –∑–∞–∫–∞–∑–∞—Ö –≤ –≥—Ä—É–ø–ø–µ B

plt.figure(figsize=(18, 4))
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ —Ä–∞–∑–º–µ—Ä–∞ –≥—Ä–∞—Ñ–∏–∫–∞

plt.plot(cumulative_data_a['date'], cumulative_data_a['conversion'], label='A')
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –ø–æ –≥—Ä—É–ø–ø–µ A

plt.plot(cumulative_data_b['date'], cumulative_data_b['conversion'], label='B')
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –ø–æ –≥—Ä—É–ø–ø–µ B

plt.legend()
# –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ª–µ–≥–µ–Ω–¥—ã –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ

plt.title('–ö—É–º—É–ª—è—Ç–∏–≤–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –ø–æ –≥—Ä—É–ø–ø–∞–º –∏ –ø–æ –¥–Ω—è–º', fontsize=12)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞

plt.xlabel('–î–∞—Ç–∞', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞

plt.ylabel('–ö—É–º—É–ª—è—Ç–∏–≤–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª-–≤–æ –∑–∞–∫–∞–∑–æ–≤', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞

plt.axis([dt.datetime(2019, 8, 1), dt.datetime(2019, 8, 31), 0, 0.05]);
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –º–∞—Å—à—Ç–∞–±–∞ –æ—Å–µ–π
# –ø–µ—Ä–µ–¥–∞—á–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –∏ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –æ—Å–∏
```

<div class="alert alert-info"> <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞: </b>

–í –Ω–∞—á–∞–ª–µ —Ç–µ—Å—Ç–∞ –≥—Ä—É–ø–ø–∞ A –∏–º–µ–ª–∞ –±–æ–ª—å—à–µ–µ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è, –Ω–æ –≤—Å–∫–æ—Ä–µ –≥—Ä—É–ø–ø–∞ B –≤—ã—Ä–æ–≤–Ω—è–ª–∞—Å—å –∏ –∑–∞—Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–ª–∞—Å—å –ø—Ä–∏–º–µ—Ä–Ω–æ –æ–∫–æ–ª–æ –æ–¥–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è <strike> (0.03) </strike> (0.033-0.035), –±–æ–ª—å—à–µ –∑–Ω–∞—á–µ–Ω–∏—è –≥—Ä—É–ø–ø—ã A <strike> (0.025) </strike> (0.029-0.03).

<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>

<b>–û—Ç–ª–∏—á–Ω–æ!‚úÖ:</b> –£–º–Ω–∏—Ü–∞, –ø—Ä–∞–≤–∏–ª—å–Ω–æ!
</div>

–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A.

```python
merged_cumulative_conversions = cumulative_data_a[['date', 'conversion']].merge(
    cumulative_data_b[['date', 'conversion']], left_on='date', right_on='date', how='left', suffixes=['_a', '_b'])
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã ‚Äòcumulative_data_a‚Äô –∏ ‚Äòcumulative_data_b‚Äô —Ç–∞–∫, 
# —á—Ç–æ–±—ã –æ–Ω–∞ —Å–æ–¥–µ—Ä–∂–∞–ª–∞ —Å—Ç–æ–ª–±—Ü—ã ‚Äòdate‚Äô, ‚Äòconversion_a‚Äô, ‚Äòconversion_b‚Äô
# —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äòmerged_cumulative_conversions‚Äô

plt.figure(figsize=(18, 4))
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ —Ä–∞–∑–º–µ—Ä–∞ –≥—Ä–∞—Ñ–∏–∫–∞

plt.plot(merged_cumulative_conversions['date'], 
         merged_cumulative_conversions['conversion_b'] / merged_cumulative_conversions['conversion_a'] - 1)
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A

plt.title('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A –ø–æ –¥–Ω—è–º', 
          fontsize=12)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞

plt.xlabel('–î–∞—Ç–∞', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞

plt.ylabel('–†–∞–∑–ª–∏—á–∏–µ –º–µ—Ç—Ä–∏–∫–∏ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞

plt.axhline(y=0, color='black', linestyle='--')
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –ø—É–Ω–∫—Ç–∏—Ä–Ω–æ–π –ª–∏–Ω–∏–∏ –æ—Å–∏ X —á—ë—Ä–Ω–æ–≥–æ —Ü–≤–µ—Ç–∞ —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º –æ—Å–∏ Y = 0

plt.axhline(y=0.15, color='grey', linestyle='--')
# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –ø—É–Ω–∫—Ç–∏—Ä–Ω–æ–π –ª–∏–Ω–∏–∏ –æ—Å–∏ X —Å–µ—Ä–æ–≥–æ —Ü–≤–µ—Ç–∞ —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º –æ—Å–∏ Y = 0.15

plt.axis([dt.datetime(2019, 8, 1), dt.datetime(2019, 8, 31), -0.3, 0.3]);
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –º–∞—Å—à—Ç–∞–±–∞ –æ—Å–µ–π
```

### –ê–Ω–∞–ª–∏–∑ –≤—ã–±—Ä–æ—Å–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö

–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏, —Å–æ–≤–µ—Ä—à–∏–≤—à–∏–µ –º–Ω–æ–≥–æ –∑–∞–∫–∞–∑–æ–≤, –≤–ª–∏—è—é—Ç –Ω–∞ –∫–æ–Ω–≤–µ—Ä—Å–∏—é. –°–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, –∏—Ö –ø–æ–≤–µ–¥–µ–Ω–∏–µ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ. –ï—Å–ª–∏ —Ä–µ—á—å –Ω–µ –∏–¥—ë—Ç –æ–± –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-–º–∞–≥–∞–∑–∏–Ω–µ —Å —Ä–µ–≥—É–ª—è—Ä–Ω—ã–º —Å–ø—Ä–æ—Å–æ–º, ¬´–æ–±—ã—á–Ω—ã–π¬ª –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å–æ–≤–µ—Ä—à–∞–µ—Ç –Ω–µ –±–æ–ª—å—à–µ –¥–≤—É—Ö –∑–∞–∫–∞–∑–æ–≤ –∑–∞ –∫–æ—Ä–æ—Ç–∫–∏–π —Å—Ä–æ–∫. –ü–æ–¥—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –ø–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º.

```python
orders_by_visitors = (orders.groupby('visitor_id', as_index=False).agg({'transaction_id': 'nunique'}))
# –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ —Å—Ç–æ–ª–±—Ü—É ‚Äòvisitor_id‚Äô
# –ø–æ–¥—Å—á—ë—Ç —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòtransaction_id‚Äô –ø–æ —Å–≥—Ä—É–ø–∏—Ä–æ–≤–∞–Ω–Ω—ã–º –¥–∞–Ω–Ω—ã–º
# —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äòorders_by_visitors‚Äô

orders_by_visitors.columns = ['visitor_id', 'orders']
# –ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞–º —Ç–∞–±–ª–∏—Ü—ã ‚Äòorders_by_visitors‚Äô

orders_by_visitors.sort_values(by='orders', ascending=False).head(15)
# —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –∑–Ω–∞—á–µ–Ω–∏–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòorders_by_visitors‚Äô –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
# –≤—ã–≤–æ–¥ –Ω–∞ —ç–∫—Ä–∞–Ω –ø–µ—Ä–≤—ã—Ö 15-—Ç–∏ —Å—Ç—Ä–æ–∫
```

### –ü–æ–¥—Å—á—ë—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –∏ –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º

–ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º ‚Äî –±–µ–∑ —É–¥–∞–ª–µ–Ω–∏—è –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—ã.

- –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –Ω–µ—Ç. 
- –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Ä–∞–∑–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –µ—Å—Ç—å.

–°–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ `orders_by_visitors_a` –∏ `orders_by_visitors_b` —Å–æ —Å—Ç–æ–ª–±—Ü–∞–º–∏ `visitor_id`, `orders`. –í –Ω–∏—Ö –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –∑–∞–∫–∞–∑—ã–≤–∞–ª–∏ —Ö–æ—Ç—è –±—ã 1 —Ä–∞–∑, —É–∫–∞–∂–µ–º —á–∏—Å–ª–æ —Å–æ–≤–µ—Ä—à—ë–Ω–Ω—ã—Ö –∑–∞–∫–∞–∑–æ–≤.

<div class="alert alert-success">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>

<b>–û—Ç–ª–∏—á–Ω–æ!‚úÖ:</b> –ì–∏–ø–æ—Ç–µ–∑—ã —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω—ã –≤–µ—Ä–Ω–æ!
</div>

```python
orders_by_visitors_a = (
    orders[orders['group'] == 'A'].groupby('visitor_id', as_index=False).agg({'transaction_id': pd.Series.nunique}))
# –≤—ã–±–æ—Ä –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–∞ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòorders‚Äô, –≥–¥–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ ‚Äògroup‚Äô —Ä–∞–≤–Ω–æ ‚ÄòA‚Äô
# –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ —Å—Ç–æ–ª–±—Ü—É ‚Äòvisitor_id‚Äô
# –ø–æ–¥—Å—á—ë—Ç —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòtransaction_id‚Äô –ø–æ —Å–≥—Ä—É–ø–∏—Ä–æ–≤–∞–Ω–Ω—ã–º –¥–∞–Ω–Ω—ã–º
# —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äòorders_by_visitors_a‚Äô

orders_by_visitors_a.columns = ['visitor_id', 'orders']
# –ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞–º —Ç–∞–±–ª–∏—Ü—ã ‚Äòorders_by_visitors_a‚Äô

orders_by_visitors_b = (
    orders[orders['group'] == 'B'].groupby('visitor_id', as_index=False).agg({'transaction_id': pd.Series.nunique}))
# –≤—ã–±–æ—Ä –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–∞ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòorders‚Äô, –≥–¥–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ ‚Äògroup‚Äô —Ä–∞–≤–Ω–æ ‚ÄòB‚Äô
# –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ —Å—Ç–æ–ª–±—Ü—É ‚Äòvisitor_id‚Äô
# –ø–æ–¥—Å—á—ë—Ç —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòtransaction_id‚Äô –ø–æ —Å–≥—Ä—É–ø–∏—Ä–æ–≤–∞–Ω–Ω—ã–º –¥–∞–Ω–Ω—ã–º
# —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äòorders_by_visitors_b‚Äô

orders_by_visitors_b.columns = ['visitor_id', 'orders']
# –ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞–º —Ç–∞–±–ª–∏—Ü—ã ‚Äòorders_by_visitors_b‚Äô
```

–û–±—ä—è–≤–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ `sample_a` –∏ `sample_b`, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º –∏–∑ —Ä–∞–∑–Ω—ã—Ö –≥—Ä—É–ø–ø –±—É–¥–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤. –¢–µ–º, –∫—Ç–æ –Ω–∏—á–µ–≥–æ –Ω–µ –∑–∞–∫–∞–∑–∞–ª, –±—É–¥—É—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –Ω—É–ª–∏. –≠—Ç–æ –Ω—É–∂–Ω–æ, —á—Ç–æ–±—ã –ø–æ–¥–≥–æ—Ç–æ–≤–∏—Ç—å –≤—ã–±–æ—Ä–∫–∏ –∫ –ø—Ä–æ–≤–µ—Ä–∫–µ –∫—Ä–∏—Ç–µ—Ä–∏–µ–º –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏.

```python
visitors_a_daily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
# —Å–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòvisitors_a_daily‚Äô, —Å–æ–¥–µ—Ä–∂–∞—â–µ–≥–æ —Ç–æ–ª—å–∫–æ —Ç–µ —Å—Ç—Ä–æ–∫–∏ –∏–∑ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòvisitors‚Äô, 
# –≥–¥–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ group —Ä–∞–≤–Ω–æ ‚ÄòA‚Äô
# –≤—ã–±–æ—Ä —Å—Ç–æ–ª–±—Ü–æ–≤ ‚Äòdate‚Äô –∏ ‚Äòvisitors‚Äô, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥—É—Ç —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ –Ω–æ–≤–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
# –ø–æ–ª—É—á–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –≤—ã–±—Ä–∞–Ω–Ω—É—é –¥–∞—Ç—É –≤ –≥—Ä—É–ø–ø–µ A

visitors_a_daily.columns = ['date', 'visitors_per_date_a']
# –ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòvisitors_a_daily‚Äô

sample_a = pd.concat(
    [orders_by_visitors_a['orders'], pd.Series(
        0, index=np.arange(
            visitors_a_daily['visitors_per_date_a'].sum() - len(orders_by_visitors_a['orders'])), name='orders')], axis=0)
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ 2-—Ö –æ–±—ä–µ–∫—Ç–æ–≤ Series: –∫–æ–ª-–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≥—Ä—É–ø–ø—ã A –∏ –Ω—É–ª–µ–π –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≥—Ä—É–ø–ø—ã A, 
# –∫–æ—Ç–æ—Ä—ã–µ –Ω–∏—á–µ–≥–æ –Ω–µ –∑–∞–∫–∞–∑—ã–≤–∞–ª–∏
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω—É–ª–µ–π —Ä–∞–≤–Ω–æ —Ä–∞–∑–Ω–∏—Ü–µ –º–µ–∂–¥—É —Å—É–º–º–æ–π –ø–æ—Å–µ—Ç–∏—Ç–µ–ª–µ–π –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–ø–∏—Å–µ–π –æ –∑–∞–∫–∞–∑–∞—Ö
# –ø–µ—Ä–µ–¥–∞—á–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∞, —Å–æ–æ–±—â–∞—é—â–µ–≥–æ, —á—Ç–æ –æ–±—ä–µ–∫—Ç—ã Series –Ω—É–∂–Ω–æ –æ–±—ä–µ–¥–∏–Ω–∏—Ç—å –ø–æ —Å—Ç—Ä–æ–∫–∞–º (—Ç–æ –µ—Å—Ç—å –∑–∞–ø–∏—Å–∞—Ç—å –ø–æ–¥—Ä—è–¥)

visitors_b_daily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
# —Å–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòvisitors_b_daily‚Äô, —Å–æ–¥–µ—Ä–∂–∞—â–µ–≥–æ —Ç–æ–ª—å–∫–æ —Ç–µ —Å—Ç—Ä–æ–∫–∏ –∏–∑ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòvisitors‚Äô, 
# –≥–¥–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ group —Ä–∞–≤–Ω–æ ‚ÄòB‚Äô
# –≤—ã–±–æ—Ä —Å—Ç–æ–ª–±—Ü–æ–≤ ‚Äòdate‚Äô –∏ ‚Äòvisitors‚Äô, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥—É—Ç —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ –Ω–æ–≤–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
# –ø–æ–ª—É—á–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –≤—ã–±—Ä–∞–Ω–Ω—É—é –¥–∞—Ç—É –≤ –≥—Ä—É–ø–ø–µ B

visitors_b_daily.columns = ['date', 'visitors_per_date_b']
# –ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòvisitors_b_daily‚Äô

sample_b = pd.concat(
    [orders_by_visitors_b['orders'], pd.Series(
        0, index=np.arange(
            visitors_b_daily['visitors_per_date_b'].sum() - len(orders_by_visitors_b['orders'])), name='orders')], axis=0)
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ 2-—Ö –æ–±—ä–µ–∫—Ç–æ–≤ Series: –∫–æ–ª-–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≥—Ä—É–ø–ø—ã B –∏ –Ω—É–ª–µ–π –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≥—Ä—É–ø–ø—ã B, 
# –∫–æ—Ç–æ—Ä—ã–µ –Ω–∏—á–µ–≥–æ –Ω–µ –∑–∞–∫–∞–∑—ã–≤–∞–ª–∏
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω—É–ª–µ–π —Ä–∞–≤–Ω–æ —Ä–∞–∑–Ω–∏—Ü–µ –º–µ–∂–¥—É —Å—É–º–º–æ–π –ø–æ—Å–µ—Ç–∏—Ç–µ–ª–µ–π –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–ø–∏—Å–µ–π –æ –∑–∞–∫–∞–∑–∞—Ö
# –ø–µ—Ä–µ–¥–∞—á–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∞, —Å–æ–æ–±—â–∞—é—â–µ–≥–æ, —á—Ç–æ –æ–±—ä–µ–∫—Ç—ã Series –Ω—É–∂–Ω–æ –æ–±—ä–µ–¥–∏–Ω–∏—Ç—å –ø–æ —Å—Ç—Ä–æ–∫–∞–º (—Ç–æ –µ—Å—Ç—å –∑–∞–ø–∏—Å–∞—Ç—å –ø–æ–¥—Ä—è–¥)
```

–ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º. –ü—Ä–∏–º–µ–Ω–∏–º —Ç–µ—Å—Ç –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏. –í—ã–≤–µ–¥–µ–º p-value –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏. –ë—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ 0.05 –≤ –∫–∞—á–µ—Å—Ç–≤–µ —É—Ä–æ–≤–Ω—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏.

–ö—Ä–æ–º–µ —Ç–æ–≥–æ, –≤—ã—á–∏—Å–ª–∏–º –∏ –≤—ã–≤–µ–¥–µ–º —Å—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã A –∏ –≥—Ä—É–ø–ø—ã B, –∞ —Ç–∞–∫–∂–µ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ —Ä–∞–∑–ª–∏—á–∏–µ –≤ —Å—Ä–µ–¥–Ω–µ–º –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ A/B-—Ç–µ—Å—Ç–∞ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º.

```python
alpha = 0.05
# —Å–æ–∑–¥–∞–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äòalpha‚Äô, —Ä–∞–≤–Ω–æ–π 0.05, –¥–ª—è —É—Å—Ç–∞–Ω–æ–≤–∫–∏ —É—Ä–æ–≤–Ω—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

pvalue = stats.mannwhitneyu(sample_a, sample_b)[1]
# –ø–æ–¥—Å—á—ë—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Ä–∞–∑–ª–∏—á–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏
# –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ç–µ—Å—Ç–∞ –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏

print(f'p-value: {"{0:.3f}".format(pvalue)}')
# –≤—ã–≤–æ–¥ p-value –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ 
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ p-value –¥–æ 3-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

if pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
# –ø—Ä–æ–≤–µ—Ä–∫–∞, –º–µ–Ω—å—à–µ –ª–∏ p-–∑–Ω–∞—á–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è, —á–µ–º —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
# –µ—Å–ª–∏ —ç—Ç–æ —Ç–∞–∫, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è –∏ –≤—ã–≤–æ–¥–∏—Ç—Å—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ
# –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤—ã–≤–æ–¥–∏—Ç—Å—è —Å—Å–æ–±—â–µ–Ω–∏–µ –æ —Ç–æ–º, —á—Ç–æ –Ω–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É

print('–°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã –ê –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º:', "{:.3f}".format(sample_a.mean()))
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –≤—ã–±–æ—Ä–∫–µ ‚Äòsample_a‚Äô
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –¥–æ 3-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

print('–°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã B –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º:', "{:.3f}".format(sample_b.mean()))
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –≤—ã–±–æ—Ä–∫–µ ‚Äòsample_b‚Äô
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –¥–æ 3-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ —Ä–∞–∑–ª–∏—á–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª-–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º:', 
      "{0:.2%}".format(sample_b.mean() / sample_a.mean() - 1))
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏ –≤—ã–≤–æ–¥ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞–∑–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–µ–º –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –¥–æ 2-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
```

–ß—Ç–æ–±—ã —Ä–∞—Å—Å—á–∏—Ç–∞—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ, –ø–µ—Ä–µ–¥–∞–¥–∏–º –∫—Ä–∏—Ç–µ—Ä–∏—é –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏ –¥–∞–Ω–Ω—ã–µ –æ –≤—ã—Ä—É—á–∫–µ —Å –∑–∞–∫–∞–∑–æ–≤. –ë—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ 0.05 –≤ –∫–∞—á–µ—Å—Ç–≤–µ —É—Ä–æ–≤–Ω—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏.

–ê –µ—â—ë –Ω–∞–π–¥—ë–º —Å—Ä–µ–¥–Ω–∏–µ —á–µ–∫–∏ –≥—Ä—É–ø–ø A –∏ B, –∞ —Ç–∞–∫–∂–µ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ —Ä–∞–∑–ª–∏—á–∏–µ –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ A/B-—Ç–µ—Å—Ç–∞ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º.

```python
pvalue = stats.mannwhitneyu(orders[orders['group'] == 'A']['revenue'], orders[orders['group'] == 'B']['revenue'])[1]
# –ø–æ–¥—Å—á—ë—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Ä–∞–∑–ª–∏—á–∏—è —Å—Ä–µ–¥–Ω–∏—Ö —á–µ–∫–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏
# –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ç–µ—Å—Ç–∞ –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏

print(f'p-value: {"{0:.3f}".format(pvalue)}')
# –≤—ã–≤–æ–¥ p-value –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–∏—Ö —á–µ–∫–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ 
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ p-value –¥–æ 3-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

if pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
# –ø—Ä–æ–≤–µ—Ä–∫–∞, –º–µ–Ω—å—à–µ –ª–∏ p-–∑–Ω–∞—á–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è, —á–µ–º —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
# –µ—Å–ª–∏ —ç—Ç–æ —Ç–∞–∫, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è –∏ –≤—ã–≤–æ–¥–∏—Ç—Å—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ
# –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤—ã–≤–æ–¥–∏—Ç—Å—è —Å—Å–æ–±—â–µ–Ω–∏–µ –æ —Ç–æ–º, —á—Ç–æ –Ω–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É

print('–°—Ä–µ–¥–Ω–∏–π —á–µ–∫ –∑–∞–∫–∞–∑–∞ –≥—Ä—É–ø–ø—ã A –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º:', 
      "{0:.2f}".format(orders[orders['group'] == 'A']['revenue'].mean()))
# –≤—ã–≤–æ–¥ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã A
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –¥–æ 2-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

print('–°—Ä–µ–¥–Ω–∏–π —á–µ–∫ –∑–∞–∫–∞–∑–∞ –≥—Ä—É–ø–ø—ã B –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º:',
      "{0:.2f}".format(orders[orders['group'] == 'B']['revenue'].mean()))
# –≤—ã–≤–æ–¥ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã B
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –¥–æ 2-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ —Ä–∞–∑–ª–∏—á–∏–µ –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ A/B-—Ç–µ—Å—Ç–∞ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º:',
    "{0:.2%}".format(
        orders[orders['group'] == 'B']['revenue'].mean() / orders[orders['group'] == 'A']['revenue'].mean() - 1))
# –≤—ã–≤–æ–¥ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –ø—Ä–∏—Ä–æ—Å—Ç–∞ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã B
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –¥–æ 2-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
```

### –ü–æ–¥—Å—á—ë—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –∏ –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º

<div class="alert alert-info"> <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞: </b>

–ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –∏ –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º.

–ù–∞–π–¥–µ–Ω–Ω—ã–µ —Ä–∞–Ω–µ–µ 95-–π –∏ 99-–π –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª–∏ —Å—Ä–µ–¥–Ω–∏—Ö —á–µ–∫–æ–≤ —Ä–∞–≤–Ω—ã <strike> 26 —Ç—ã—Å. 785 –∏ 53 —Ç—ã—Å. 904 </strike> 28 —Ç—ã—Å. –∏ 58 —Ç—ã—Å. 233. –ê 95-–π –∏ 99-–π –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è —Ä–∞–≤–Ω—ã <strike> –æ–¥–Ω–æ–º—É –∏ –¥–≤—É–º –∑–∞–∫–∞–∑–∞–º </strike> –¥–≤—É–º –∏–ª–∏ —á–µ—Ç—ã—Ä—ë–º –∑–∞–∫–∞–∑–∞–º –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.

–ú—ã –ø—Ä–∏–Ω—è–ª–∏ –∑–∞ –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Ç–µ—Ö, –∫—Ç–æ —Å–æ–≤–µ—Ä—à–∏–ª 3 –∑–∞–∫–∞–∑–∞ –∏–ª–∏ —Å–æ–≤–µ—Ä—à–∏–ª –∑–∞–∫–∞–∑ –¥–æ—Ä–æ–∂–µ 30 —Ç—ã—Å. –¢–∞–∫ –º—ã —É–±–µ—Ä—ë–º <strike> 1% </strike> 5% –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –∑–∞–∫–∞–∑–æ–≤ –∏ –æ—Ç 1% –¥–æ 5% –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –¥–æ—Ä–æ–≥–∏–º–∏ –∑–∞–∫–∞–∑–∞–º–∏.

–°–¥–µ–ª–∞–µ–º —Å—Ä–µ–∑—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å —á–∏—Å–ª–æ–º –∑–∞–∫–∞–∑–æ–≤ –±–æ–ª—å—à–µ 2 ‚Äî `visitors_with_many_orders` –∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, —Å–æ–≤–µ—Ä—à–∏–≤—à–∏—Ö –∑–∞–∫–∞–∑—ã –¥–æ—Ä–æ–∂–µ 30 000 ‚Äî `visitors_with_expensive_orders`. –û–±—ä–µ–¥–∏–Ω–∏–º –∏—Ö –≤ —Ç–∞–±–ª–∏—Ü–µ `abnormal_visitors`.

```python
# number_of_orders_limit = np.percentile(orders_by_visitors['orders'], 99)
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ 99-–≥–æ –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòorders‚Äô –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ ‚Äòorders_by_visitors‚Äô
# –∑–∞–¥–∞–Ω–∏–µ –≥—Ä–∞–Ω–∏—Ü—ã –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤

number_of_orders_limit = np.percentile(orders_by_visitors['orders'], 95)
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ 95-–≥–æ –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòorders‚Äô –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ ‚Äòorders_by_visitors‚Äô
# –∑–∞–¥–∞–Ω–∏–µ –≥—Ä–∞–Ω–∏—Ü—ã –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤

visitors_with_many_orders = pd.concat(
    [orders_by_visitors_a[orders_by_visitors_a['orders'] > number_of_orders_limit]['visitor_id'],
        orders_by_visitors_b[orders_by_visitors_b['orders'] > number_of_orders_limit]['visitor_id'],],
    axis=0)
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª–µ–π, 
# —É –∫–æ—Ç–æ—Ä—ã—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –ø—Ä–µ–≤—ã—à–∞–µ—Ç –∑–∞–¥–∞–Ω–Ω–æ–µ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ

# order_value_limit = round(np.percentile(orders['revenue'], 95.905))
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ 95.905-–≥–æ –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòrevenue‚Äô –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ ‚Äòorders‚Äô
# –∑–∞–¥–∞–Ω–∏–µ –≥—Ä–∞–Ω–∏—Ü—ã —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –∑–∞–∫–∞–∑–∞

order_value_limit = round(np.percentile(orders['revenue'], 95.5214))
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ 95.5214-–≥–æ –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ ‚Äòrevenue‚Äô –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ ‚Äòorders‚Äô
# –∑–∞–¥–∞–Ω–∏–µ –≥—Ä–∞–Ω–∏—Ü—ã —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –∑–∞–∫–∞–∑–∞

visitors_with_expensive_orders = orders[orders['revenue'] > order_value_limit]['visitor_id']
# –æ—Ç–±–æ—Ä –ø–æ—Å–µ—Ç–∏—Ç–µ–ª–µ–π, —É –∫–æ—Ç–æ—Ä—ã—Ö –µ—Å—Ç—å –¥–æ—Ä–æ–≥–∏–µ –∑–∞–∫–∞–∑—ã (–≥–¥–µ –≤—ã—Ä—É—á–∫–∞ –±–æ–ª—å—à–µ –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è)

abnormal_visitors = (pd.concat([visitors_with_many_orders, visitors_with_expensive_orders], axis=0)
    .drop_duplicates()
    .sort_values())
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –≤ —Ç–∞–±–ª–∏—Ü–µ ‚Äòabnormal_visitors‚Äô –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å —á–∏—Å–ª–æ–º –∑–∞–∫–∞–∑–æ–≤ –±–æ–ª—å—à–µ 2 
# –∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, —Å–æ–≤–µ—Ä—à–∏–≤—à–∏—Ö –∑–∞–∫–∞–∑—ã –¥–æ—Ä–æ–∂–µ 30 000
# —É–¥–∞–ª–µ–Ω–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤, —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é

abnormal_visitors.head(5).to_frame()
# –≤—ã–≤–æ–¥ –ø–µ—Ä–≤—ã—Ö 5-—Ç–∏ —Å—Ç—Ä–æ–∫ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äòabnormal_visitors‚Äô
```

<div class="alert alert-info"> <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞: </b>

–í—Å–µ–≥–æ <strike> 49 </strike> 68 –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π. –£–∑–Ω–∞–µ–º, –∫–∞–∫ –∏—Ö –¥–µ–π—Å—Ç–≤–∏—è –ø–æ–≤–ª–∏—è–ª–∏ –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ç–µ—Å—Ç–∞. –ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ —Ç–µ—Å—Ç–∞ –ø–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º ‚Äî –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

- –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –Ω–µ—Ç.
- –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Ä–∞–∑–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –µ—Å—Ç—å.

–°–Ω–∞—á–∞–ª–∞ –ø–æ–¥–≥–æ—Ç–æ–≤–∏–º –≤—ã–±–æ—Ä–∫–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –ø–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º –ø–æ –≥—Ä—É–ø–ø–∞–º A/B-—Ç–µ—Å—Ç–∞.
    
</div>

```python
sample_a_filtered = pd.concat(
    [
        orders_by_visitors_a[
            np.logical_not(orders_by_visitors_a['visitor_id'].isin(abnormal_visitors))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitors_a_daily['visitors_per_date_a'].sum() - len(orders_by_visitors_a['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0)
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ 2-—Ö –æ–±—ä–µ–∫—Ç–æ–≤ Series: –∫–æ–ª-–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≥—Ä—É–ø–ø—ã A –∏ –Ω—É–ª–µ–π –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≥—Ä—É–ø–ø—ã A, 
# –∫–æ—Ç–æ—Ä—ã–µ –Ω–∏—á–µ–≥–æ –Ω–µ –∑–∞–∫–∞–∑—ã–≤–∞–ª–∏ 
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω—É–ª–µ–π —Ä–∞–≤–Ω–æ —Ä–∞–∑–Ω–∏—Ü–µ –º–µ–∂–¥—É —Å—É–º–º–æ–π –ø–æ—Å–µ—Ç–∏—Ç–µ–ª–µ–π –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–ø–∏—Å–µ–π –æ –∑–∞–∫–∞–∑–∞—Ö
# —Å –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π

sample_b_filtered = pd.concat(
    [
        orders_by_visitors_b[
            np.logical_not(orders_by_visitors_b['visitor_id'].isin(abnormal_visitors))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitors_b_daily['visitors_per_date_b'].sum() - len(orders_by_visitors_b['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0)
# –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ 2-—Ö –æ–±—ä–µ–∫—Ç–æ–≤ Series: –∫–æ–ª-–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≥—Ä—É–ø–ø—ã B –∏ –Ω—É–ª–µ–π –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≥—Ä—É–ø–ø—ã B, 
# –∫–æ—Ç–æ—Ä—ã–µ –Ω–∏—á–µ–≥–æ –Ω–µ –∑–∞–∫–∞–∑—ã–≤–∞–ª–∏ 
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω—É–ª–µ–π —Ä–∞–≤–Ω–æ —Ä–∞–∑–Ω–∏—Ü–µ –º–µ–∂–¥—É —Å—É–º–º–æ–π –ø–æ—Å–µ—Ç–∏—Ç–µ–ª–µ–π –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–ø–∏—Å–µ–π –æ –∑–∞–∫–∞–∑–∞—Ö
# —Å –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
```

–ü—Ä–∏–º–µ–Ω–∏–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏ –∫ –ø–æ–ª—É—á–µ–Ω–Ω—ã–º –≤—ã–±–æ—Ä–∫–∞–º `sample_a_filtered` –∏ `sample_b_filtered`.

–ë—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ 0.05 –≤ –∫–∞—á–µ—Å—Ç–≤–µ —É—Ä–æ–≤–Ω—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏.

```python
pvalue = stats.mannwhitneyu(sample_a_filtered, sample_b_filtered)[1]
# –ø–æ–¥—Å—á—ë—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Ä–∞–∑–ª–∏—á–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏
# –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ç–µ—Å—Ç–∞ –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏

print(f'p-value: {"{0:.3f}".format(pvalue)}')
# –≤—ã–≤–æ–¥ p-value –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ 
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ p-value –¥–æ 3-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

if pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
# –ø—Ä–æ–≤–µ—Ä–∫–∞, –º–µ–Ω—å—à–µ –ª–∏ p-–∑–Ω–∞—á–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è, —á–µ–º —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
# –µ—Å–ª–∏ —ç—Ç–æ —Ç–∞–∫, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è –∏ –≤—ã–≤–æ–¥–∏—Ç—Å—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ
# –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤—ã–≤–æ–¥–∏—Ç—Å—è —Å—Å–æ–±—â–µ–Ω–∏–µ –æ —Ç–æ–º, —á—Ç–æ –Ω–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É

print('–°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã –ê –ø–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º:', "{:.3f}".format(sample_a_filtered.mean()))
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –≤—ã–±–æ—Ä–∫–µ ‚Äòsample_a_filtered‚Äô
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –¥–æ 3-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

print('–°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã B –ø–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º:', "{:.3f}".format(sample_b_filtered.mean()))
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –≤—ã–±–æ—Ä–∫–µ ‚Äòsample_b_filtered‚Äô
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –¥–æ 3-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ —Ä–∞–∑–ª–∏—á–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª-–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º:', 
      "{0:.2%}".format(sample_b_filtered.mean() / sample_a_filtered.mean() - 1))
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏ –≤—ã–≤–æ–¥ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞–∑–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–µ–º –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏
# –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –¥–æ 2-—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
```
