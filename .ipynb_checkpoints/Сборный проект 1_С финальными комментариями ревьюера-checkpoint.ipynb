–î–ª—è –ª—É—á—à–µ–π –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏–∏ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã–π —Ç–∏–ø –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö —Å –≥–æ–¥–∞–º–∏ –≤—ã–ø—É—Å–∫–∞ –∏–≥—Ä –∏ –æ—Ü–µ–Ω–∫–∞–º–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤, —Ç–∞–∫ –∫–∞–∫ –æ–Ω–∏ —Å–æ–¥–µ—Ä–∂–∞—Ç –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞, —É –∫–æ—Ç–æ—Ä—ã—Ö –¥—Ä–æ–±–Ω–∞—è —á–∞—Å—Ç—å —Ä–∞–≤–Ω–∞ –Ω—É–ª—é.

### –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å—Ç–æ–ª–±—Ü–∞ —Å –æ–±—â–∏–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –≤–æ –≤—Å–µ—Ö —Ä–µ–≥–∏–æ–Ω–∞—Ö

–ü–æ—Å—á–∏—Ç–∞–µ–º –æ–±—â–∏–µ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –≤–æ –≤—Å–µ—Ö —Ä–µ–≥–∏–æ–Ω–∞—Ö (–≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ, –ï–≤—Ä–æ–ø–µ, –Ø–ø–æ–Ω–∏–∏ –∏ –¥—Ä—É–≥–∏—Ö —Å—Ç—Ä–∞–Ω–∞—Ö) –∏ –∑–∞–ø–∏—à–µ–º –∏—Ö –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü.

```python
data['total_sales'] = data[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
# —Å–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ ‚Äòtotal_sales‚Äô, —Ä–∞–≤–Ω–æ–≥–æ —Å—É–º–º–µ –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤ ‚Äòna_sales‚Äô, ‚Äòeu_sales‚Äô, ‚Äòjp_sales‚Äô –∏ ‚Äòother_sales‚Äô
data['total_sales'].head().to_frame()
# –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ø–µ—Ä–≤—ã—Ö 5 —Å—Ç—Ä–æ–∫ —Å—Ç–æ–ª–±—Ü–∞ ‚Äòtotal_sales‚Äô
```

–í –¥–∞—Ç–∞—Ñ—Ä–µ–π–º `data` –¥–æ–±–∞–≤–ª–µ–Ω —Å—Ç–æ–ª–±–µ—Ü —Å –æ–±—â–∏–º –æ–±—ä—ë–º–æ–º –ø—Ä–æ–¥–∞–∂ –∫–∞–∂–¥–æ–π –∏–≥—Ä—ã.
    
–í—ã–≤–µ–¥–µ–º –ø–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ `data` –Ω–∞ —ç–∫—Ä–∞–Ω –ø–æ—Å–ª–µ –≤–Ω–µ—Å–µ–Ω–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏–π.

```python
data.head() 
# –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ø–µ—Ä–≤—ã—Ö 5 —Å—Ç—Ä–æ–∫ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòdata‚Äô
```

–í—ã–≤–µ–¥–µ–º –æ—Å–Ω–æ–≤–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ `data` –ø–æ—Å–ª–µ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö.

```python
data.info() 
# –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Å—Ç—Ä—É–∫—Ç—É—Ä–µ –¥–∞–Ω–Ω—ã—Ö, —Ö—Ä–∞–Ω—è—â–∏—Ö—Å—è –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äòdata‚Äô
```

–í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö –≤ —Ç–∞–±–ª–∏—Ü–µ —Å—Ç–∞–ª–æ 16 —Ç—ã—Å. 443 —Å—Ç—Ä–æ–∫–∏ –∏ 12 —Å—Ç–æ–ª–±—Ü–æ–≤. –®–µ—Å—Ç—å —Å—Ç–æ–ª–±—Ü–æ–≤ —Å–æ–¥–µ—Ä–∂–∞—Ç –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ (`'na_sales'`, `'eu_sales'`, `'jp_sales'`, `'other_sales'`, `'user_score'`, `'total_sales'`), —á–µ—Ç—ã—Ä–µ ‚Äî —Å—Ç—Ä–æ–∫–∏ (`'name'`, `'platform'`, `'genre'`, `'rating'`), –¥–≤–∞ ‚Äî —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ (`'year_of_release'`, `'critic_score'`). –í–æ –≤—Å–µ—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.

### –í—ã–≤–æ–¥

–ü—Ä–∏–≤–µ–ª–∏ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É –Ω–∞–∑–≤–∞–Ω–∏—è –≤—Å–µ—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ `data`.

–£–¥–∞–ª–∏–ª–∏ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö —Å –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ –∏–≥—Ä, –∂–∞–Ω—Ä–∞–º–∏ –∏ –≥–æ–¥–∞–º–∏ –≤—ã–ø—É—Å–∫–∞.

–í —Å—Ç–æ–ª–±—Ü–µ —Å –æ—Ü–µ–Ω–∫–∞–º–∏ –∏–≥—Ä –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏ –æ–±–Ω–∞—Ä—É–∂–∏–ª–∏ –∑–Ω–∞—á–µ–Ω–∏—è 'tbd', —É–∫–∞–∑—ã–≤–∞—é—â–∏–µ –Ω–∞ —Ç–æ, —á—Ç–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –µ—â—ë –Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞. –ò–∑–º–µ–Ω–∏–ª–∏ —ç—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—â–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è (NaN). –ó–∞—Ç–µ–º –∑–∞–º–µ–Ω–∏–ª–∏ –∫–∞–∂–¥—ã–π –ø—Ä–æ–ø—É—Å–∫ –Ω–∞ –æ—Ü–µ–Ω–∫—É, —è–≤–ª—è—é—â—É—é—Å—è –≤—ã—à–µ —Ç–æ–π, –∫–æ—Ç–æ—Ä—É—é –±—ã –º–æ–≥–ª–∏ –ø–æ—Å—Ç–∞–≤–∏—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏.

–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ —Å —Ä–µ–π—Ç–∏–Ω–≥–∞–º–∏ –∏–≥—Ä –æ—Ç –æ—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏–∏ ESRB –∑–∞–º–µ–Ω–∏–ª–∏ –Ω–∞ —Å–ª–æ–≤–æ 'Undetermined'.

–ó–∞–ø–æ–ª–Ω–∏–ª–∏ –∫–∞–∂–¥–æ–µ –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ —Å—Ç–æ–ª–±—Ü–µ, —Å–æ–¥–µ—Ä–∂–∞—â–∏–º –æ—Ü–µ–Ω–∫–∏ –∏–≥—Ä –∫—Ä–∏—Ç–∏–∫–∞–º–∏, –æ—Ü–µ–Ω–∫–æ–π, —è–≤–ª—è—é—â–µ–π—Å—è –≤—ã—à–µ —Ç–æ–π, –∫–æ—Ç–æ—Ä—É—é –±—ã –º–æ–≥–ª–∏ –ø–æ—Å—Ç–∞–≤–∏—Ç—å –∫—Ä–∏—Ç–∏–∫–∏.

–í —Å—Ç–æ–ª–±—Ü–∞—Ö —Å –≥–æ–¥–∞–º–∏ –≤—ã–ø—É—Å–∫–∞ –∏–≥—Ä –∏ –æ—Ü–µ–Ω–∫–∞–º–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–ª–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –≤ —Ü–µ–ª—ã–µ.

–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–ª–∏ —Å—Ç—Ä–æ–∫–∏ –≤ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –≤ —Å—Ç–æ–ª–±—Ü–µ, –≥–¥–µ —É–∫–∞–∑–∞–Ω—ã –æ—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.
    
–û–±–Ω–∞—Ä—É–∂–∏–ª–∏ –∏ —É–¥–∞–ª–∏–ª–∏ –æ–¥–∏–Ω –¥—É–±–ª–∏–∫–∞—Ç –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—é –∏–≥—Ä—ã, –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ –∏ –≥–æ–¥—É –≤—ã–ø—É—Å–∫–∞.

–î–æ–±–∞–≤–∏–ª–∏ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º `data` —Å—Ç–æ–ª–±–µ—Ü —Å –æ–±—â–∏–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –≤–æ –≤—Å–µ—Ö —Ä–µ–≥–∏–æ–Ω–∞—Ö.

<font color='DarkBlue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞</b></font><br>
<font color='DarkGreen'>üëå –•–æ—Ä–æ—à–æ, —á—Ç–æ —Å–¥–µ–ª–∞–Ω—ã –≤—ã–≤–æ–¥—ã –ø–æ —Ä–∞–∑–¥–µ–ª—É –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö. üëç</font>

```python
# –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞
# –ü–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –æ—Å—Ç–∞–ª–æ—Å—å
temp = data.copy() 
list_c = ['name', 'platform', 'year_of_release', 'genre', 'critic_score', 'user_score', 'rating']
print(temp.info())
for col_l in list_c:
  print('-'* 25)
  print(col_l, temp[col_l].sort_values().unique())
  print(col_l,': –∫–æ–ª-–≤–æ NaN',temp[col_l].isna().sum(),
        ', –ø—Ä–æ—Ü–µ–Ω—Ç NaN', round(temp[col_l].isna().mean()*100,2),'%') 
```

## –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

### –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä –≤ —Ä–∞–Ω—ã–µ –≥–æ–¥—ã –≤—ã–ø—É—Å–∫–∞

–û–ø—Ä–µ–¥–µ–ª–∏–º, —Å–∫–æ–ª—å–∫–æ –∏–≥—Ä –≤—ã–ø—É—Å–∫–∞–ª–æ—Å—å –≤ —Ä–∞–∑–Ω—ã–µ –≥–æ–¥—ã.

```python
number_of_games_by_year = data.pivot_table(index='year_of_release', values='name', aggfunc='count')
# —Å–æ–∑–¥–∞–Ω–∏–µ —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòdata‚Äô
# –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ —Å—Ç–æ–ª–±—Ü—É ‚Äòyear_of_release‚Äô
# –≤–æ–∑–≤—Ä–∞—â–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤—Ö–æ–∂–¥–µ–Ω–∏–π –≤ —Ç–∞–±–ª–∏—Ü—É –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ ‚Äòname‚Äô
number_of_games_by_year.columns = ['number_of_games']
# c–æ–∑–¥–∞–Ω–∏–µ —Å–ø–∏—Å–∫–∞ —Å –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è —Ç–∞–±–ª–∏—Ü—ã ‚Äònumber_of_games_by_year‚Äô, –∫–æ—Ç–æ—Ä—ã–π –ø–µ—Ä–µ–¥–∞—ë—Ç—Å—è –≤ –∞—Ç—Ä–∏–±—É—Ç ‚Äòcolumns‚Äô
sns.lineplot(data=number_of_games_by_year, marker='o', legend=False).grid(5)
# —Å–æ–∑–¥–∞–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Å –≥–æ–¥–∞–º–∏ —Ä–µ–ª–∏–∑–∞ –Ω–∞ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –æ—Å–∏ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∏–≥—Ä –Ω–∞ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –æ—Å–∏
sns.set_style('whitegrid')
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ —Å—Ç–∏–ª—è –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
plt.title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–ø—É—â–µ–Ω–Ω—ã—Ö –∏–≥—Ä –≤ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–º –≥–æ–¥—É', fontsize=12)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
plt.xlabel('–ì–æ–¥ –≤—ã–ø—É—Å–∫–∞', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä', fontsize=10)
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –æ—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞
number_of_games_by_year
# –≤—ã–≤–æ–¥ —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã
```

```python
current_data = data.query('year_of_release > 2013')
# —Å–æ–∑–¥–∞–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äòcurrent_data‚Äô, –≤–∫–ª—é—á–∞—é—â–µ–π –≤—Å–µ –∑–∞–ø–∏—Å–∏ –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòdata‚Äô, 
# —É –∫–æ—Ç–æ—Ä—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–µ –≤ —Å—Ç–æ–ª–±—Ü–µ ‚Äòyear_of_release‚Äô –±–æ–ª—å—à–µ 2013
len(current_data)
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòcurrent_data‚Äô
```

### –î–∏–Ω–∞–º–∏–∫–∞ –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö –∑–∞ –∞–∫—Ç—É–∞–ª—å–Ω—ã–π –ø–µ—Ä–∏–æ–¥

```python
total_sales_by_platform_and_year = current_data.pivot_table(index=['platform', 'year_of_release'],
                                           values='total_sales', 
                                           aggfunc='sum')
# —Å–æ–∑–¥–∞–Ω–∏–µ —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòcurrent_data‚Äô
# –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º ‚Äòplatform‚Äô –∏ ‚Äòyear_of_release‚Äô
# —Å—É–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ ‚Äòtotal_sales‚Äô
total_sales_by_platform_and_year
# –≤—ã–≤–æ–¥ —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã
```

```python
def sales(sale):
    def platform_categorization(platform):
        if platform in list((current_data.pivot_table(index='platform', 
                                  values=sale, 
                                  aggfunc='sum').sort_values(sale, 
                                                             ascending=False)).index[0:5]):
            return platform
        else:
            return 'other'
    sales_on_platforms = current_data.pivot_table(index=current_data['platform'].apply(platform_categorization), 
                         values=sale, 
                         aggfunc='sum').sort_values(sale, 
                                                    ascending=False)
    return sales_on_platforms
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ ‚Äòsales‚Äô
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ ‚Äòplatform_categorization‚Äô
# –ø—Ä–æ–≤–µ—Ä–∫–∞, –≤—Ö–æ–¥–∏—Ç –ª–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ –≤ —Ç–æ–ø-5 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º
# –µ—Å–ª–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ –Ω–∞–π–¥–µ–Ω–∞ –≤ —Å–ø–∏—Å–∫–µ, –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç—Å—è —ç—Ç–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞
# –µ—Å–ª–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ –Ω–µ –Ω–∞–π–¥–µ–Ω–∞ –≤ —Å–ø–∏—Å–∫–µ, –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ ‚Äòother‚Äô
# —Å–æ–∑–¥–∞–Ω–∏–µ —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ ‚Äòcurrent_data‚Äô
# –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –ø–æ ‚Äòplatform_categorization‚Äô
# —Å—É–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ —Å –ø—Ä–æ–¥–∞–∂–∞–º–∏
# —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —É–±—ã–≤–∞–Ω–∏—é
colors = sns.color_palette('twilight_shifted')
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–≤–µ—Ç–æ–≤–æ–π –ø–∞–ª–∏—Ç—Ä—ã
def pie_charts(title): # –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ ‚Äòpie_charts‚Äô
    fig, axs = plt.subplots(1, 3, figsize=(14, 9))
    axs[0].pie(data=sales('na_sales'),
                         x='na_sales',
                         labels=sales('na_sales').index[:],
                         autopct='%1.0f%%',
                         wedgeprops={'edgecolor':'black'},
                         colors=colors)
    axs[1].pie(data=sales('eu_sales'),
                         x='eu_sales',
                         labels=sales('eu_sales').index[:],
                         autopct='%1.0f%%',
                         wedgeprops={'edgecolor':'black'},
                         colors=colors)
    axs[2].pie(data=sales('jp_sales'),
                         x='jp_sales',
                         labels=sales('jp_sales').index[:],
                         autopct='%1.0f%%',
                         wedgeprops={'edgecolor':'black'},
                         colors=colors)
    fig.suptitle(f'–î–æ–ª–∏ –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç {title} –≤ —Ä–∞–∑–Ω—ã—Ö —Ä–µ–≥–∏–æ–Ω–∞—Ö, %')
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
    plt.subplots_adjust(top=1.35);
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –º–µ–∂–¥—É –æ–±—â–∏–º –∑–∞–≥–æ–ª–æ–≤–∫–æ–º –∏ –ø–æ–¥–∑–∞–≥–æ–ª–æ–≤–∫–∞–º–∏
    axs[0].set_title('–°–µ–≤–µ—Ä–Ω–∞—è –ê–º–µ—Ä–∏–∫–∞')
    axs[1].set_title('–ï–≤—Ä–æ–ø–∞')
    axs[2].set_title('–Ø–ø–æ–Ω–∏—è')
# —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–æ–¥–∑–∞–≥–æ–ª–æ–≤–∫–æ–≤
    plt.show()
pie_charts('–ø–ª–∞—Ç—Ñ–æ—Ä–º—ã')
# —Å–æ–∑–¥–∞–Ω–∏–µ –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞ –ø—Ä–æ–¥–∞–∂–∏
```

–í –Ø–ø–æ–Ω–∏–∏ –±–æ–ª–µ–µ –ø–æ–ª–æ–≤–∏–Ω—ã –ø—Ä–æ–¥–∞–∂ (56%) –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ –æ—Ç–µ—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–º –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º –æ—Ç ¬´Nintendo¬ª. 48% –ø—Ä–æ–¥–∞–∂ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—Ç –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ 3DS, 8% ‚Äî Wii U. –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –æ–¥–∏–Ω–∞–∫–æ–≤–æ –Ω–∞ —è–ø–æ–Ω—Å–∫–æ–º —Ä—ã–Ω–∫–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∞ —Å–µ—Ä–∏—è –∏–≥—Ä–æ–≤—ã—Ö –∫–æ–Ω—Å–æ–ª–µ–π PlayStation –æ—Ç ¬´Sony¬ª ‚Äî PlayStation 4 (16%), PlayStation Vita (16%) –∏ PlayStation 3 (12%). –õ–∏—à—å 1% –ø—Ä–æ–¥–∞–∂ —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç –æ—Å—Ç–∞–ª—å–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã.

```python
alpha = 0.01
# —Å–æ–∑–¥–∞–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äòalpha‚Äô, —Ä–∞–≤–Ω–æ–π 0.01, –¥–ª—è —É—Å—Ç–∞–Ω–æ–≤–∫–∏ —É—Ä–æ–≤–Ω—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

user_score_xbox_one = current_data.query('platform == "XOne" and user_score < 10.1')['user_score']
# –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏ –∏–≥—Ä –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ Xbox One

user_score_pc = current_data.query('platform == "PC" and user_score < 10.1')['user_score']
# –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏ –∏–≥—Ä –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ PC

results = st.ttest_ind(user_score_xbox_one, user_score_pc, equal_var=False)
# –ø—Ä–æ–≤–µ–¥–µ–Ω–∏–µ ttest –¥–ª—è –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫

print(f'p-value: {results.pvalue}')
# –≤—ã–≤–æ–¥ –Ω–∞ —ç–∫—Ä–∞–Ω –ø–æ–ª—É—á–∏–≤—à–µ–≥–æ—Å—è p-–∑–Ω–∞—á–µ–Ω–∏—è

if results.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
# –ø—Ä–æ–≤–µ—Ä–∫–∞, –º–µ–Ω—å—à–µ –ª–∏ p-–∑–Ω–∞—á–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è, —á–µ–º —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 
# –µ—Å–ª–∏ —ç—Ç–æ —Ç–∞–∫, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è –∏ –≤—ã–≤–æ–¥–∏—Ç—Å—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ
# –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤—ã–≤–æ–¥–∏—Ç—Å—è —Å–æ–æ–±—â–µ–Ω–∏–µ –æ —Ç–æ–º, —á—Ç–æ –Ω–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–∞—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É
```

–ù–µ–ª—å–∑—è –æ–¥–Ω–æ–∑–Ω–∞—á–Ω–æ —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ –ª—É—á—à–µ ‚Äî –ø—Ä–∏—Å—Ç–∞–≤–∫–∞ Xbox One –∏–ª–∏ –ø–µ—Ä—Å–æ–Ω–∞–ª—å–Ω—ã–π –∫–æ–º–ø—å—é—Ç–µ—Ä, —Ç–∞–∫ –∫–∞–∫ –æ–±–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –∏–º–µ—é—Ç —Å–∏–ª—å–Ω—ã–µ –∏ —Å–ª–∞–±—ã–µ —Å—Ç–æ—Ä–æ–Ω—ã –≤ —Ç–µ—Ö–Ω–∏—á–µ—Å–∫–æ–º –ø–ª–∞–Ω–µ –∏ —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è –¥–æ—Å—Ç—É–ø–Ω–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –∏–≥—Ä. –ù–∞ Xbox One –ª—É—á—à–µ –∏–≥—Ä–∞—Ç—å, –Ω–∞–ø—Ä–∏–º–µ—Ä, –≤ —Ñ–∞–π—Ç–∏–Ω–≥–∏, –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ—Ä—ã, —Å–ø–æ—Ä—Ç–∏–≤–Ω—ã–µ –∏–≥—Ä—ã. –í –∏–≥—Ä–∞—Ö —Å –∞–∫—Ç–∏–≤–Ω—ã–º–∏ –¥–≤–∏–∂–µ–Ω–∏—è–º–∏ –ø–µ—Ä—Å–æ–Ω–∞–∂–∞ –≥–µ–π–º–ø–∞–¥ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –Ω–µ–∑–∞–º–µ–Ω–∏–º—ã–º. –î–ª—è —à—É—Ç–µ—Ä–æ–≤ –∏ —Å—Ç—Ä–∞—Ç–µ–≥–∏–π –±–æ–ª—å—à–µ –ø–æ–¥—Ö–æ–¥–∏—Ç –∫–ª–∞–≤–∏–∞—Ç—É—Ä–∞. –ù–∞ –ø–µ—Ä—Å–æ–Ω–∞–ª—å–Ω—ã—Ö –∫–æ–º–ø—å—é—Ç–µ—Ä–∞—Ö –±–æ–ª—å—à–µ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–µ–π –¥–ª—è –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –≥—Ä–∞—Ñ–∏–∫–∏ –∏ —É–ø—Ä–∞–≤–ª–µ–Ω–∏—è.

### –ì–∏–ø–æ—Ç–µ–∑–∞: ¬´—Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports —Ä–∞–∑–Ω—ã–µ¬ª

–ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action (–∞–Ω–≥–ª. ¬´–¥–µ–π—Å—Ç–≤–∏–µ¬ª, —ç–∫—à–µ–Ω-–∏–≥—Ä—ã) –∏ Sports (–∞–Ω–≥–ª. ¬´—Å–ø–æ—Ä—Ç–∏–≤–Ω—ã–µ —Å–æ—Ä–µ–≤–Ω–æ–≤–∞–Ω–∏—è¬ª) —Ä–∞–∑–Ω—ã–µ. –†–∞—Å—Å—á–∏—Ç–∞–µ–º —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–æ–∫ –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è, –ø–æ –∫–æ—Ç–æ—Ä—ã–º –º–æ–∂–Ω–æ —Å—É–¥–∏—Ç—å –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ –¥–∏—Å–ø–µ—Ä—Å–∏–π —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π.

```python
round(
    current_data.query(
        '(genre in ["Action", "Sports"]) and user_score < 10.1').groupby(
        'genre')['user_score'].describe(), 1)
# –≤—ã–≤–æ–¥ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –∑–Ω–∞—á–µ–Ω–∏—è—Ö —Å—Ç–æ–ª–±—Ü–∞ ‚Äòuser_score‚Äô –¥–ª—è –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports
```

–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫ –∏–≥—Ä –≤ –∂–∞–Ω—Ä–µ —ç–∫—à–µ–Ω (297) –∏ —Å–ø–æ—Ä—Ç–∏–≤–Ω—ã—Ö –∏–≥—Ä (127) —Å–∏–ª—å–Ω–æ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è. –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –æ—Ü–µ–Ω–æ–∫ –∂–∞–Ω—Ä–æ–≤ Action (1.4) –∏ Sports (1.9) –Ω–µ —Å–æ–≤–ø–∞–¥–∞—é—Ç, —á—Ç–æ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ —Ç–æ, —á—Ç–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Ç–æ–∂–µ –Ω–µ —è–≤–ª—è—é—Ç—Å—è –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º–∏. –£–∫–∞–∂–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä equal_var –∫–∞–∫ False –ø—Ä–∏ –≤—ã–∑–æ–≤–µ –º–µ—Ç–æ–¥–∞ scipy.stats.ttest_ind(), —á—Ç–æ–±—ã —Ç–µ—Å—Ç –ø–æ—Å—á–∏—Ç–∞–ª –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫ –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞.

–û–ø—Ä–µ–¥–µ–ª–∏–º, —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è –ª–∏ —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports.

- –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports —Ä–∞–≤–Ω—ã.
- –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports –Ω–µ —Ä–∞–≤–Ω—ã.
    
–£—Å—Ç–∞–Ω–æ–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ 0.01 (1%) –≤ –∫–∞—á–µ—Å—Ç–≤–µ —É—Ä–æ–≤–Ω—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏. –≠—Ç–æ—Ç —É—Ä–æ–≤–µ–Ω—å –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ–±–Ω–∞—Ä—É–∂–∏—Ç—å –∏—Å—Ç–∏–Ω–Ω—ã–µ —Ä–∞–∑–ª–∏—á–∏—è —Å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç—å—é –∏ –∏–∑–±–µ–∂–∞—Ç—å –æ—à–∏–±–æ—á–Ω—ã—Ö –≤—ã–≤–æ–¥–æ–≤.

```python
alpha = 0.01
# —Å–æ–∑–¥–∞–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äòalpha‚Äô, —Ä–∞–≤–Ω–æ–π 0.01, –¥–ª—è —É—Å—Ç–∞–Ω–æ–≤–∫–∏ —É—Ä–æ–≤–Ω—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

user_score_action = current_data.query('genre == "Action" and user_score < 10.1')['user_score']
# –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏ –∏–≥—Ä –≤ –∂–∞–Ω—Ä–µ action

user_score_sports = current_data.query('genre == "Sports" and user_score < 10.1')['user_score']
# –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏ –∏–≥—Ä –≤ –∂–∞–Ω—Ä–µ sports

results = st.ttest_ind(user_score_action, user_score_sports, equal_var=False)
# –ø—Ä–æ–≤–µ–¥–µ–Ω–∏–µ ttest –¥–ª—è –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫

print(f'p-value: {results.pvalue}')
# –≤—ã–≤–æ–¥ –Ω–∞ —ç–∫—Ä–∞–Ω –ø–æ–ª—É—á–∏–≤—à–µ–≥–æ—Å—è p-–∑–Ω–∞—á–µ–Ω–∏—è

if results.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
# –ø—Ä–æ–≤–µ—Ä–∫–∞, –º–µ–Ω—å—à–µ –ª–∏ p-–∑–Ω–∞—á–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è, —á–µ–º —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 
# –µ—Å–ª–∏ —ç—Ç–æ —Ç–∞–∫, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è –∏ –≤—ã–≤–æ–¥–∏—Ç—Å—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ
# –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤—ã–≤–æ–¥–∏—Ç—Å—è —Å–æ–æ–±—â–µ–Ω–∏–µ –æ —Ç–æ–º, —á—Ç–æ –Ω–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–∞—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É
```
